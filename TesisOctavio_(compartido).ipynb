```python
from __future__ import annotations

from dataclasses import dataclass
from typing import Dict,Any
import pandas as pd
from CoolProp.CoolProp import PropsSI

@dataclass
class Corriente:
    n:int
    f:str
    P:float  # kPa
    T:float  # °C
    m:float  # kg/s
    h:float  # kJ/kg
    s:float  # kJ/kgK
    Ex:float  # kW
    Ex_fisica:float  # kW
    Ex_quimica:float  # kW

def _T_desde_hP_incomp(f,P_Pa,h_obj_Jkg,T_hi_K):
    Tlo=285.15  # K
    hlo=PropsSI("H","T",Tlo,"P",P_Pa,f)
    if h_obj_Jkg<hlo:
        raise ValueError("h_obj<h(Tlo,P)")
    Thi=T_hi_K  # K
    for _ in range(80):
        try:
            hhi=PropsSI("H","T",Thi,"P",P_Pa,f)
        except Exception:
            Thi=0.98*Thi
            continue
        if hhi>=h_obj_Jkg:
            break
        Thi=1.02*Thi
    a,b=Tlo,Thi
    for _ in range(80):
        Tm=0.5*(a+b)  # K
        hm=PropsSI("H","T",Tm,"P",P_Pa,f)
        if hm>=h_obj_Jkg:
            b=Tm
        else:
            a=Tm
    return 0.5*(a+b)  # K

def _ex_fis(m,h,s,h0,s0,T0_K):
    ex_Jkg=(h-h0)-T0_K*(s-s0)
    return m*ex_Jkg/1000.0  # kW

def ejecutar_ptc(param:Dict[str,Any])->Dict[int,Corriente]:
    T0=param["T0"]  # °C
    P0=param["P0"]  # kPa
    f=param["fluido_ptc"]  # -
    m=param["m_th"]  # kg/s
    P1=param["P1"]  # kPa
    P2=param["P2"]  # kPa
    T1=param["T1"]  # °C
    Q=param["Qsolar_util"]  # kW

    T0K=T0+273.15  # K
    P0Pa=P0*1000.0  # Pa
    P1Pa=P1*1000.0  # Pa
    P2Pa=P2*1000.0  # Pa
    T1K=T1+273.15  # K

    h0=PropsSI("H","T",T0K,"P",P0Pa,f)
    s0=PropsSI("S","T",T0K,"P",P0Pa,f)
    h1=PropsSI("H","T",T1K,"P",P1Pa,f)
    s1=PropsSI("S","T",T1K,"P",P1Pa,f)

    dh=Q*1000.0/m  # J/kg
    h2=h1-dh
    T2K=_T_desde_hP_incomp(f,P2Pa,h2,670.15)  # K
    s2=PropsSI("S","T",T2K,"P",P2Pa,f)

    Ex1_fis=_ex_fis(m,h1,s1,h0,s0,T0K)  # kW
    Ex2_fis=_ex_fis(m,h2,s2,h0,s0,T0K)  # kW

    Ex1_quim=0.0  # kW
    Ex2_quim=0.0  # kW

    Ex1=Ex1_fis+Ex1_quim  # kW
    Ex2=Ex2_fis+Ex2_quim  # kW

    c={}
    c[1]=Corriente(1,f,P1,T1,m,h1/1000.0,s1/1000.0,Ex1,Ex1_fis,Ex1_quim)
    c[2]=Corriente(2,f,P2,T2K-273.15,m,h2/1000.0,s2/1000.0,Ex2,Ex2_fis,Ex2_quim)

    df=pd.DataFrame([{
        "Nodo":v.n,
        "Fluido":v.f,
        "P(kPa)":v.P,
        "T(ºC)":v.T,
        "m(kg/s)":v.m,
        "h(kJ/kg)":v.h,
        "s(kJ/kg K)":v.s,
        "Ex(kW)":v.Ex,
        "Ex_fisica(kW)":v.Ex_fisica,
        "Ex_quimica(kW)":v.Ex_quimica
    } for v in c.values()])
    print(df.to_string(index=False))

    param["T2"]=T2K-273.15  # °C
    param["h2"]=h2/1000.0  # kJ/kg
    param["s2"]=s2/1000.0  # kJ/kgK
    return c

corr_ptc=ejecutar_ptc(parametros_base)
```

```python
import pandas as pd
from CoolProp.CoolProp import PropsSI

def _air_TP(T,P):
    TK=T+273.15  # K
    PP=P*1000.0  # Pa
    h=PropsSI("H","T",TK,"P",PP,"Air")/1000.0  # kJ/kg
    s=PropsSI("S","T",TK,"P",PP,"Air")/1000.0  # kJ/kgK
    return h,s

def _air_PS(P,s):
    PP=P*1000.0  # Pa
    ss=s*1000.0  # J/kgK
    h=PropsSI("H","P",PP,"S",ss,"Air")/1000.0  # kJ/kg
    return h

def _air_PH(P,h):
    PP=P*1000.0  # Pa
    hh=h*1000.0  # J/kg
    TK=PropsSI("T","P",PP,"H",hh,"Air")  # K
    s=PropsSI("S","P",PP,"H",hh,"Air")/1000.0  # kJ/kgK
    return TK-273.15,s

def _w_TP(T,P):
    TK=T+273.15  # K
    PP=P*1000.0  # Pa
    h=PropsSI("H","T",TK,"P",PP,"Water")/1000.0  # kJ/kg
    s=PropsSI("S","T",TK,"P",PP,"Water")/1000.0  # kJ/kgK
    return h,s

def _w_Px(P,x):
    PP=P*1000.0  # Pa
    h=PropsSI("H","P",PP,"Q",x,"Water")/1000.0  # kJ/kg
    s=PropsSI("S","P",PP,"Q",x,"Water")/1000.0  # kJ/kgK
    T=PropsSI("T","P",PP,"Q",x,"Water")-273.15  # °C
    return T,h,s

def _w_PS(P,s):
    PP=P*1000.0  # Pa
    ss=s*1000.0  # J/kgK
    h=PropsSI("H","P",PP,"S",ss,"Water")/1000.0  # kJ/kg
    return h

def _w_PH(P,h):
    PP=P*1000.0  # Pa
    hh=h*1000.0  # J/kg
    T=PropsSI("T","P",PP,"H",hh,"Water")-273.15  # °C
    s=PropsSI("S","P",PP,"H",hh,"Water")/1000.0  # kJ/kgK
    return T,s

def _exfis(m,h,s,h0,s0,T0):
    T0K=T0+273.15  # K
    ex=(h-h0)-T0K*(s-s0)  # kJ/kg
    return m*ex  # kW

def ejecutar_efgt_A(p):
    T0=p["T0"]  # °C
    P0=p["P0"]  # kPa

    T11=p["T11"]  # °C
    P11=p["P11"]  # kPa
    m=p["m_aire"]  # kg/s

    rp=p["relacion_presiones_compresor"]  # -
    etac=p["eta_compresor"]  # -
    etat=p["eta_turbina_gas"]  # -
    etag=p["eta_generador_efgt"]  # -

    T1=p["T1"]  # °C
    dTsph=p["deltaT_pp_prec_solar"]  # °C
    T14=p["T14"]  # °C
    T17=p["T17"]  # °C

    P12=P11*rp  # kPa
    P13=P12  # kPa
    P14=P12  # kPa
    P15=P11  # kPa
    P16=P11  # kPa
    P17=P11  # kPa

    h0,s0=_air_TP(T0,P0)

    h11,s11=_air_TP(T11,P11)
    Ex11_fis=_exfis(m,h11,s11,h0,s0,T0)  # kW
    Ex11_quim=0.0  # kW
    Ex11=Ex11_fis+Ex11_quim  # kW

    h12s=_air_PS(P12,s11)
    h12=h11+(h12s-h11)/etac
    T12,s12=_air_PH(P12,h12)
    Ex12_fis=_exfis(m,h12,s12,h0,s0,T0)  # kW
    Ex12_quim=0.0  # kW
    Ex12=Ex12_fis+Ex12_quim  # kW

    T13=T1-dTsph  # °C
    h13,s13=_air_TP(T13,P13)
    Ex13_fis=_exfis(m,h13,s13,h0,s0,T0)  # kW
    Ex13_quim=0.0  # kW
    Ex13=Ex13_fis+Ex13_quim  # kW

    h14,s14=_air_TP(T14,P14)
    Ex14_fis=_exfis(m,h14,s14,h0,s0,T0)  # kW
    Ex14_quim=0.0  # kW
    Ex14=Ex14_fis+Ex14_quim  # kW

    h15s=_air_PS(P15,s14)
    h15=h14-etat*(h14-h15s)
    T15,s15=_air_PH(P15,h15)
    Ex15_fis=_exfis(m,h15,s15,h0,s0,T0)  # kW
    Ex15_quim=0.0  # kW
    Ex15=Ex15_fis+Ex15_quim  # kW

    h17,s17=_air_TP(T17,P17)
    Ex17_fis=_exfis(m,h17,s17,h0,s0,T0)  # kW
    Ex17_quim=0.0  # kW
    Ex17=Ex17_fis+Ex17_quim  # kW

    Qsph=m*(h13-h12)  # kW
    Qhthe=m*(h14-h13)  # kW

    P21=p["P21"]  # kPa
    Pcond=p["P_cond_src"]  # kPa
    etats=p["eta_turbina_src"]  # -
    etap=p["eta_bomba_src"]  # -
    etags=p["eta_generador_src"]  # -
    dTsh=p["deltaT_pp_sobrecalentador_src"]  # °C

    Wsrc=p.get("Wsrc_obj",p.get("W_src_obj",p.get("W_src_obj_kW",p.get("W_src_obj_kW",p.get("W_src",p.get("W_src_kW",p.get("W_src_obj_kW",p.get("W_src_obj",111.2))))))))  # kW

    T21=T15-dTsh  # °C
    h21,s21=_w_TP(T21,P21)

    T23,h23,s23=_w_Px(Pcond,0.0)

    h24s=_w_PS(P21,s23)
    h24=h23+(h24s-h23)/etap
    T24,s24=_w_PH(P21,h24)

    h22s=_w_PS(Pcond,s21)
    h22=h21-etats*(h21-h22s)

    wts=h21-h22  # kJ/kg
    wps=h24-h23  # kJ/kg
    wnet=max(1e-12,(wts-wps))  # kJ/kg
    mele=max(1e-12,etags*wnet)  # kJ/kg

    m21=Wsrc/mele  # kg/s
    Qsg=m21*(h21-h24)  # kW

    h16=h15-Qsg/m
    T16,s16=_air_PH(P16,h16)
    Ex16_fis=_exfis(m,h16,s16,h0,s0,T0)  # kW
    Ex16_quim=0.0  # kW
    Ex16=Ex16_fis+Ex16_quim  # kW

    Wc=m*(h12-h11)  # kW
    Wt=m*(h14-h15)  # kW
    Wm=Wt-Wc  # kW
    We=etag*Wm  # kW

    Q1516=m*(h16-h15)  # kW
    Q1617=m*(h17-h16)  # kW
    Q1517=m*(h17-h15)  # kW

    c={}
    c[11]=Corriente(11,"aire",P11,T11,m,h11,s11,Ex11,Ex11_fis,Ex11_quim)
    c[12]=Corriente(12,"aire",P12,T12,m,h12,s12,Ex12,Ex12_fis,Ex12_quim)
    c[13]=Corriente(13,"aire",P13,T13,m,h13,s13,Ex13,Ex13_fis,Ex13_quim)
    c[14]=Corriente(14,"aire",P14,T14,m,h14,s14,Ex14,Ex14_fis,Ex14_quim)
    c[15]=Corriente(15,"aire",P15,T15,m,h15,s15,Ex15,Ex15_fis,Ex15_quim)
    c[16]=Corriente(16,"aire",P16,T16,m,h16,s16,Ex16,Ex16_fis,Ex16_quim)
    c[17]=Corriente(17,"aire",P17,T17,m,h17,s17,Ex17,Ex17_fis,Ex17_quim)

    df=pd.DataFrame([{
        "Nodo":c[k].n,
        "Fluido":c[k].f,
        "P(kPa)":c[k].P,
        "T(ºC)":c[k].T,
        "m(kg/s)":c[k].m,
        "h(kJ/kg)":c[k].h,
        "s(kJ/kg K)":c[k].s,
        "Ex(kW)":c[k].Ex,
        "Ex_fisica(kW)":c[k].Ex_fisica,
        "Ex_quimica(kW)":c[k].Ex_quimica
    } for k in [11,12,13,14,15,16,17]])
    print(df.to_string(index=False))

    p["P12"]=P12  # kPa
    p["T12"]=T12  # °C
    p["T13"]=T13  # °C
    p["T15"]=T15  # °C
    p["T16"]=T16  # °C

    p["Wcomp"]=Wc  # kW
    p["Wgt"]=Wt  # kW
    p["Wmec"]=Wm  # kW
    p["Welec"]=We  # kW

    p["Qsph"]=Qsph  # kW
    p["Qhthe"]=Qhthe  # kW
    p["Q15_16"]=Q1516  # kW
    p["Q16_17"]=Q1617  # kW
    p["Q15_17"]=Q1517  # kW

    p["T21_est"]=T21  # °C
    p["m21_est"]=m21  # kg/s
    p["Qsg_est"]=Qsg  # kW
    p["h21_est"]=h21  # kJ/kg
    p["h24_est"]=h24  # kJ/kg
    p["h22_est"]=h22  # kJ/kg
    return c

corr_efgt=ejecutar_efgt_A(parametros_base)
```

```python
import pandas as pd
import cantera as ct

def _normX(X):
    s=sum(max(v,0.0) for v in X.values())
    if s<=0:
        raise ValueError("X_invalida")
    return {k:max(v,0.0)/s for k,v in X.items() if v>0.0}

def _Xstr(X):
    X=_normX(X)
    return ",".join([f"{k}:{v:.12g}" for k,v in X.items()])

def _hsMW(g,T,P,X):
    TK=T+273.15  # K
    PP=P*1000.0  # Pa
    g.TPX=TK,PP,_Xstr(X)
    h=g.enthalpy_mass/1000.0  # kJ/kg
    s=g.entropy_mass/1000.0  # kJ/kgK
    MW=g.mean_molecular_weight  # kg/kmol
    return h,s,MW

def _T_hP(g,P,X,hobj,Tmin,Tmax):
    a=Tmin
    b=Tmax
    ha=_hsMW(g,a,P,X)[0]
    hb=_hsMW(g,b,P,X)[0]
    if (ha-hobj)*(hb-hobj)>0:
        if hobj<min(ha,hb)-1e-6 or hobj>max(ha,hb)+1e-6:
            raise ValueError("invertir_h_fallo")
        return a if abs(ha-hobj)<=abs(hb-hobj) else b
    for _ in range(140):
        m=0.5*(a+b)
        hm=_hsMW(g,m,P,X)[0]
        if (ha-hobj)*(hm-hobj)<=0:
            b=m
            hb=hm
        else:
            a=m
            ha=hm
    return 0.5*(a+b)

def _bag_CHO(Cw,Hw,Ow,Aw):
    C=Cw
    H=Hw
    O=Ow
    A=Aw
    if C>1.0 or H>1.0 or O>1.0 or A>1.0:
        C=C/100.0
        H=H/100.0
        O=O/100.0
        A=A/100.0
    s=C+H+O+A
    if s<=0:
        raise ValueError("bagazo_invalido")
    if abs(s-1.0)>1e-3:
        C=C/s
        H=H/s
        O=O/s
        A=A/s
    morg=1.0-A
    if morg<=0:
        raise ValueError("bagazo_invalido")
    C=C/morg
    H=H/morg
    O=O/morg
    nC=C/12.011
    nH=H/1.008
    nO=O/15.999
    if nC<=0:
        raise ValueError("bagazo_invalido")
    Cb=1.0
    Hb=nH/nC
    Ob=nO/nC
    MW=12.011*Cb+1.008*Hb+15.999*Ob  # kg/kmol
    return Cb,Hb,Ob,MW

def _prod(nbag,Cb,Hb,Ob,nch4,nO2,nN2):
    nuO2b=Cb+Hb/4.0-Ob/2.0
    if nuO2b<=0:
        raise ValueError("O2_esteq_invalido")
    nO2req=nbag*nuO2b+2.0*nch4
    if nO2+1e-12<nO2req:
        raise ValueError("O2_insuficiente")
    nCO2=nbag*Cb+nch4
    nH2O=nbag*(Hb/2.0)+2.0*nch4
    nO2exc=nO2-nO2req
    return {"CO2":nCO2,"H2O":nH2O,"O2":nO2exc,"N2":nN2}

def _exfis(m,h,s,h0,s0,T0):
    T0K=T0+273.15  # K
    ex=(h-h0)-T0K*(s-s0)  # kJ/kg
    return m*ex  # kW

def _exqfuel(m,LHV,beta):
    return m*beta*LHV  # kW

def _exqmix(g,X,X0,T0,P0):
    T0K=T0+273.15  # K
    P0P=P0*1000.0  # Pa
    X=_normX(X)
    X0=_normX(X0)
    g.TPX=T0K,P0P,_Xstr(X)
    mu=list(g.chemical_potentials)
    MW=g.mean_molecular_weight  # kg/kmol
    g.TPX=T0K,P0P,_Xstr(X0)
    mu0=list(g.chemical_potentials)
    exkmol=0.0  # J/kmol
    for sp,xi in X.items():
        i=g.species_index(sp)
        exkmol+=xi*(mu[i]-mu0[i])
    exJkg=exkmol/MW  # J/kg
    return exJkg/1000.0  # kJ/kg

def ejecutar_bcc(p):
    T0=p["T0"]  # °C
    P0=p["P0"]  # kPa
    m3=p["m3"]  # kg/s
    m4=p["m4"]  # kg/s
    T5=p["T5_max"]  # °C
    eta=p["eta_camara_combustion"]  # -
    LHVb=p["PCI_bagazo"]  # kJ/kg
    LHVg=p["PCI_gn"]  # kJ/kg
    usar_gn=bool(p.get("usar_gn_si_falta_calor",True))  # -

    if "Qhthe_efgt" in p:
        Qhthe=p["Qhthe_efgt"]  # kW
    elif "Qhthe" in p:
        Qhthe=p["Qhthe"]  # kW
        p["Qhthe_efgt"]=Qhthe  # kW
    else:
        raise KeyError("falta_Qhthe_efgt_o_Qhthe")

    T7=p.get("T7_min",100.0)  # °C

    Cw=p.get("bagazo_base_seca_C",49.5)  # % masa
    Hw=p.get("bagazo_base_seca_H",5.8)  # % masa
    Ow=p.get("bagazo_base_seca_O",43.5)  # % masa
    Aw=p.get("bagazo_base_seca_cenizas",1.2)  # % masa

    yO2=p.get("yO2",0.21)  # -
    yN2=p.get("yN2",0.79)  # -
    MWair=p.get("MWair",28.97)  # kg/kmol

    beta_b=p.get("beta_b",1.06)  # -
    beta_g=p.get("beta_g",1.06)  # -
    inc_exq_prod=bool(p.get("inc_exq_prod",True))  # -

    Xamb=dict(p.get("Xamb",{"N2":0.7808,"O2":0.2095,"CO2":0.0004,"H2O":1e-6}))
    Xair={"O2":yO2,"N2":yN2}

    if m4<=0 or eta<=0 or eta>1:
        raise ValueError("m4_o_eta_invalido")
    if m3<0:
        m3=0.0

    g=ct.Solution("gri30.yaml")

    nair=m4/MWair  # kmol/s
    nO2=nair*yO2  # kmol/s
    nN2=nair*yN2  # kmol/s

    Cb,Hb,Ob,MWb=_bag_CHO(Cw,Hw,Ow,Aw)
    nbag=m3/MWb  # kmol/s

    h4,s4,_=_hsMW(g,T0,P0,Xair)
    h0a,s0a,_=_hsMW(g,T0,P0,Xair)

    mgn=0.0  # kg/s
    X5=None

    for _ in range(70):
        nch4=mgn/16.043 if mgn>0 else 0.0  # kmol/s
        mol=_prod(nbag,Cb,Hb,Ob,nch4,nO2,nN2)
        X5=_normX(mol)
        h5,s5,_=_hsMW(g,T5,P0,X5)
        m5=m3+m4+mgn  # kg/s
        Qreq=m5*h5-m4*h4  # kW
        Qdisp=eta*(m3*LHVb+mgn*LHVg)  # kW
        falta=Qreq-Qdisp  # kW
        if falta>1e-6:
            if not usar_gn:
                raise ValueError("falta_calor")
            mgn2=mgn+falta/(eta*LHVg)  # kg/s
        else:
            mgn2=0.0
        if abs(mgn2-mgn)<=max(1e-10,1e-5*max(1.0,mgn2)):
            mgn=mgn2
            break
        mgn=mgn2

    m5=m3+m4+mgn  # kg/s
    h5,s5,_=_hsMW(g,T5,P0,X5)
    h0p,s0p,_=_hsMW(g,T0,P0,X5)

    Ex4_fisica=_exfis(m4,h4,s4,h0a,s0a,T0)  # kW
    Ex4_quimica=0.0  # kW
    Ex4=Ex4_fisica+Ex4_quimica  # kW

    ex5q=0.0  # kJ/kg
    Ex5_quimica=0.0  # kW
    if inc_exq_prod:
        ex5q=_exqmix(g,X5,Xamb,T0,P0)
        Ex5_quimica=m5*ex5q  # kW

    Ex5_fisica=_exfis(m5,h5,s5,h0p,s0p,T0)  # kW
    Ex5=Ex5_fisica+Ex5_quimica  # kW

    h6=h5-Qhthe/m5 if m5>0 else h5  # kJ/kg
    T6=_T_hP(g,P0,X5,h6,T7,T5) if m5>0 else T5  # °C
    h6,s6,_=_hsMW(g,T6,P0,X5)

    h7,s7,_=_hsMW(g,T7,P0,X5)

    Ex6_fisica=_exfis(m5,h6,s6,h0p,s0p,T0)  # kW
    Ex7_fisica=_exfis(m5,h7,s7,h0p,s0p,T0)  # kW

    Ex6_quimica=Ex5_quimica  # kW
    Ex7_quimica=Ex5_quimica  # kW

    Ex6=Ex6_fisica+Ex6_quimica  # kW
    Ex7=Ex7_fisica+Ex7_quimica  # kW

    Ex3_fisica=0.0  # kW
    Ex3_quimica=_exqfuel(m3,LHVb,beta_b)  # kW
    Ex3=Ex3_fisica+Ex3_quimica  # kW

    Exg_fisica=0.0  # kW
    Exg_quimica=_exqfuel(mgn,LHVg,beta_g) if mgn>0 else 0.0  # kW
    Exg=Exg_fisica+Exg_quimica  # kW

    c={}
    c[3]=Corriente(3,"bagazo",P0,T0,m3,float("nan"),float("nan"),Ex3,Ex3_fisica,Ex3_quimica)
    c[4]=Corriente(4,"aire",P0,T0,m4,h4,s4,Ex4,Ex4_fisica,Ex4_quimica)
    c[5]=Corriente(5,"gases",P0,T5,m5,h5,s5,Ex5,Ex5_fisica,Ex5_quimica)
    c[6]=Corriente(6,"gases",P0,T6,m5,h6,s6,Ex6,Ex6_fisica,Ex6_quimica)
    c[7]=Corriente(7,"gases",P0,T7,m5,h7,s7,Ex7,Ex7_fisica,Ex7_quimica)

    df=pd.DataFrame([{
        "Nodo":c[k].n,
        "Fluido":c[k].f,
        "P(kPa)":c[k].P,
        "T(ºC)":c[k].T,
        "m(kg/s)":c[k].m,
        "h(kJ/kg)":c[k].h,
        "s(kJ/kg K)":c[k].s,
        "Ex(kW)":c[k].Ex,
        "Ex_fisica(kW)":c[k].Ex_fisica,
        "Ex_quimica(kW)":c[k].Ex_quimica
    } for k in [3,4,5,6,7]])
    print(df.to_string(index=False))

    p["mgn"]=mgn  # kg/s
    p["X5"]=dict(X5)
    p["T6"]=T6  # °C
    p["Ex3"]=Ex3  # kW
    p["Exg"]=Exg  # kW
    p["Ex3_fisica"]=Ex3_fisica  # kW
    p["Ex3_quimica"]=Ex3_quimica  # kW
    p["Exg_fisica"]=Exg_fisica  # kW
    p["Exg_quimica"]=Exg_quimica  # kW
    p["Qhthe_efgt"]=Qhthe  # kW
    return c

corr_bcc=ejecutar_bcc(parametros_base)
```

```python
import pandas as pd
from CoolProp.CoolProp import PropsSI

def _w_TP(T,P):
    TK=T+273.15  # K
    PP=P*1000.0  # Pa
    h=PropsSI("H","T",TK,"P",PP,"Water")/1000.0  # kJ/kg
    s=PropsSI("S","T",TK,"P",PP,"Water")/1000.0  # kJ/kgK
    return h,s

def _w_Px(P,x):
    PP=P*1000.0  # Pa
    h=PropsSI("H","P",PP,"Q",x,"Water")/1000.0  # kJ/kg
    s=PropsSI("S","P",PP,"Q",x,"Water")/1000.0  # kJ/kgK
    T=PropsSI("T","P",PP,"Q",x,"Water")-273.15  # °C
    return T,h,s

def _w_PS(P,s):
    PP=P*1000.0  # Pa
    ss=s*1000.0  # J/kgK
    h=PropsSI("H","P",PP,"S",ss,"Water")/1000.0  # kJ/kg
    return h

def _w_PH(P,h):
    PP=P*1000.0  # Pa
    hh=h*1000.0  # J/kg
    T=PropsSI("T","P",PP,"H",hh,"Water")-273.15  # °C
    s=PropsSI("S","P",PP,"H",hh,"Water")/1000.0  # kJ/kgK
    return T,s

def _exfis(m,h,s,h0,s0,T0):
    T0K=T0+273.15  # K
    ex=(h-h0)-T0K*(s-s0)  # kJ/kg
    return m*ex  # kW

def ejecutar_src(p,c_efgt):
    T0=p["T0"]  # °C
    P0=p["P0"]  # kPa

    P21=p["P21"]  # kPa
    P23=p["P_cond_src"]  # kPa
    etat=p["eta_turbina_src"]  # -
    etap=p["eta_bomba_src"]  # -
    etag=p["eta_generador_src"]  # -

    dTsh=p["deltaT_pp_sobrecalentador_src"]  # °C

    if 15 not in c_efgt or 16 not in c_efgt:
        raise KeyError("faltan_15_o_16_en_efgt")
    c15=c_efgt[15]
    c16=c_efgt[16]

    ma=c15.m  # kg/s
    Qvg=ma*(c15.h-c16.h)  # kW
    if Qvg<=0:
        raise ValueError("Qvg<=0_revisa_15_16")

    T21=c15.T-dTsh  # °C
    h21,s21=_w_TP(T21,P21)

    h22s=_w_PS(P23,s21)
    h22=h21-etat*(h21-h22s)
    T22,s22=_w_PH(P23,h22)

    T23,h23,s23=_w_Px(P23,0.0)

    h24s=_w_PS(P21,s23)
    h24=h23+(h24s-h23)/etap
    T24,s24=_w_PH(P21,h24)

    den=h21-h24  # kJ/kg
    if abs(den)<1e-12:
        raise ValueError("h21_h24_muy_cercanos")
    m21=Qvg/den  # kg/s
    if m21<=0:
        raise ValueError("m21<=0")

    Wt=m21*(h21-h22)  # kW
    Wb=m21*(h24-h23)  # kW
    Wn=Wt-Wb  # kW
    We=etag*Wn  # kW

    Qin=m21*(h21-h24)  # kW
    Qc=m21*(h22-h23)  # kW

    h0,s0=_w_TP(T0,P0)

    Ex21_fisica=_exfis(m21,h21,s21,h0,s0,T0)  # kW
    Ex22_fisica=_exfis(m21,h22,s22,h0,s0,T0)  # kW
    Ex23_fisica=_exfis(m21,h23,s23,h0,s0,T0)  # kW
    Ex24_fisica=_exfis(m21,h24,s24,h0,s0,T0)  # kW

    Ex21_quimica=0.0  # kW
    Ex22_quimica=0.0  # kW
    Ex23_quimica=0.0  # kW
    Ex24_quimica=0.0  # kW

    Ex21=Ex21_fisica+Ex21_quimica  # kW
    Ex22=Ex22_fisica+Ex22_quimica  # kW
    Ex23=Ex23_fisica+Ex23_quimica  # kW
    Ex24=Ex24_fisica+Ex24_quimica  # kW

    c={}
    c[21]=Corriente(21,"agua",P21,T21,m21,h21,s21,Ex21,Ex21_fisica,Ex21_quimica)
    c[22]=Corriente(22,"agua",P23,T22,m21,h22,s22,Ex22,Ex22_fisica,Ex22_quimica)
    c[23]=Corriente(23,"agua",P23,T23,m21,h23,s23,Ex23,Ex23_fisica,Ex23_quimica)
    c[24]=Corriente(24,"agua",P21,T24,m21,h24,s24,Ex24,Ex24_fisica,Ex24_quimica)

    df=pd.DataFrame([{
        "Nodo":c[k].n,
        "Fluido":c[k].f,
        "P(kPa)":c[k].P,
        "T(ºC)":c[k].T,
        "m(kg/s)":c[k].m,
        "h(kJ/kg)":c[k].h,
        "s(kJ/kg K)":c[k].s,
        "Ex(kW)":c[k].Ex,
        "Ex_fisica(kW)":c[k].Ex_fisica,
        "Ex_quimica(kW)":c[k].Ex_quimica
    } for k in [21,22,23,24]])
    print(df.to_string(index=False))

    p["T21"]=T21  # °C
    p["T22"]=T22  # °C
    p["T23"]=T23  # °C
    p["T24"]=T24  # °C
    p["m21"]=m21  # kg/s

    p["Qvg"]=Qvg  # kW
    p["Qin_src"]=Qin  # kW
    p["Qc_src"]=Qc  # kW
    p["Wt_src"]=Wt  # kW
    p["Wb_src"]=Wb  # kW
    p["Wn_src"]=Wn  # kW
    p["We_src"]=We  # kW

    return c

corr_src=ejecutar_src(parametros_base,corr_efgt)
```

```python
import pandas as pd
from CoolProp.CoolProp import PropsSI

def _orc_TP(T,P,f):
    TK=T+273.15  # K
    PP=P*1000.0  # Pa
    h=PropsSI("H","T",TK,"P",PP,f)/1000.0  # kJ/kg
    s=PropsSI("S","T",TK,"P",PP,f)/1000.0  # kJ/kgK
    return h,s

def _orc_Px(P,x,f):
    PP=P*1000.0  # Pa
    h=PropsSI("H","P",PP,"Q",x,f)/1000.0  # kJ/kg
    s=PropsSI("S","P",PP,"Q",x,f)/1000.0  # kJ/kgK
    T=PropsSI("T","P",PP,"Q",x,f)-273.15  # °C
    return T,h,s

def _orc_Ph(P,h,f):
    PP=P*1000.0  # Pa
    hh=h*1000.0  # J/kg
    T=PropsSI("T","P",PP,"H",hh,f)-273.15  # °C
    s=PropsSI("S","P",PP,"H",hh,f)/1000.0  # kJ/kgK
    return T,s

def _w_TP(T,P):
    TK=T+273.15  # K
    PP=P*1000.0  # Pa
    h=PropsSI("H","T",TK,"P",PP,"Water")/1000.0  # kJ/kg
    s=PropsSI("S","T",TK,"P",PP,"Water")/1000.0  # kJ/kgK
    return h,s

def _w_Ph(P,h):
    PP=P*1000.0  # Pa
    hh=h*1000.0  # J/kg
    T=PropsSI("T","P",PP,"H",hh,"Water")-273.15  # °C
    s=PropsSI("S","P",PP,"H",hh,"Water")/1000.0  # kJ/kgK
    return T,s

def _exfis(m,h,s,h0,s0,T0):
    T0K=T0+273.15  # K
    ex=(h-h0)-T0K*(s-s0)  # kJ/kg
    return m*ex  # kW

def ejecutar_orc(p,c_efgt):
    T0=p["T0"]  # °C
    P0=p["P0"]  # kPa

    f=p["fluido_orc"]  # -
    P31=p["P31"]  # kPa
    T31=p["T31"]  # °C
    etat=p["eta_turbina_orc"]  # -
    etab=p["eta_bomba_orc"]  # -
    etag=p["eta_generador_orc"]  # -
    Tcond=p["T_cond_orc"]  # °C

    if 16 not in c_efgt or 17 not in c_efgt:
        raise KeyError("ORC requiere 16 y 17 del EFGT")
    c16=c_efgt[16]
    c17=c_efgt[17]

    Qdisp=c16.m*(c16.h-c17.h)  # kW
    if Qdisp<=0:
        raise ValueError("Qdisp<=0 en 16_17")

    h0o,s0o=_orc_TP(T0,P0,f)
    h0w,s0w=_w_TP(T0,P0)

    h31,s31=_orc_TP(T31,P31,f)

    TK=Tcond+273.15  # K
    P33=PropsSI("P","T",TK,"Q",0.0,f)/1000.0  # kPa
    T33,h33,s33=_orc_Px(P33,0.0,f)

    P32=P33  # kPa
    h32s=PropsSI("H","P",P32*1000.0,"S",s31*1000.0,f)/1000.0  # kJ/kg
    h32=h31-etat*(h31-h32s)  # kJ/kg
    T32,s32=_orc_Ph(P32,h32,f)

    P34=P31  # kPa
    h34s=PropsSI("H","P",P34*1000.0,"S",s33*1000.0,f)/1000.0  # kJ/kg
    h34=h33+(h34s-h33)/etab  # kJ/kg
    T34,s34=_orc_Ph(P34,h34,f)

    den=h31-h34  # kJ/kg
    if den<=0:
        raise ValueError("h31_h34<=0")

    morc=Qdisp/den  # kg/s
    if morc<=0:
        raise ValueError("morc<=0")

    WT=morc*(h31-h32)  # kW
    Wb=morc*(h34-h33)  # kW
    Wn=WT-Wb  # kW
    We=etag*Wn  # kW

    Qin=morc*(h31-h34)  # kW
    Qc=morc*(h32-h33)  # kW

    T38=p["T38"]  # °C
    P38=p["P38"]  # kPa
    T39_obj=p["T39"]  # °C
    P39=p.get("P39",P38)  # kPa

    h38,s38=_w_TP(T38,P38)

    if "m38" in p:
        m38=p["m38"]  # kg/s
        if m38<=0:
            raise ValueError("m38<=0")
        h39=h38+Qc/m38  # kJ/kg
        T39,s39=_w_Ph(P39,h39)
    else:
        h39,s39=_w_TP(T39_obj,P39)
        denw=h39-h38  # kJ/kg
        if denw<=0:
            raise ValueError("h39_h38<=0")
        m38=Qc/denw  # kg/s
        T39=T39_obj  # °C

    Ex31_fisica=_exfis(morc,h31,s31,h0o,s0o,T0)  # kW
    Ex32_fisica=_exfis(morc,h32,s32,h0o,s0o,T0)  # kW
    Ex33_fisica=_exfis(morc,h33,s33,h0o,s0o,T0)  # kW
    Ex34_fisica=_exfis(morc,h34,s34,h0o,s0o,T0)  # kW

    Ex38_fisica=_exfis(m38,h38,s38,h0w,s0w,T0)  # kW
    Ex39_fisica=_exfis(m38,h39,s39,h0w,s0w,T0)  # kW

    Ex31_quimica=0.0  # kW
    Ex32_quimica=0.0  # kW
    Ex33_quimica=0.0  # kW
    Ex34_quimica=0.0  # kW
    Ex38_quimica=0.0  # kW
    Ex39_quimica=0.0  # kW

    Ex31=Ex31_fisica+Ex31_quimica  # kW
    Ex32=Ex32_fisica+Ex32_quimica  # kW
    Ex33=Ex33_fisica+Ex33_quimica  # kW
    Ex34=Ex34_fisica+Ex34_quimica  # kW
    Ex38=Ex38_fisica+Ex38_quimica  # kW
    Ex39=Ex39_fisica+Ex39_quimica  # kW

    c={}
    c[31]=Corriente(31,f,P31,T31,morc,h31,s31,Ex31,Ex31_fisica,Ex31_quimica)
    c[32]=Corriente(32,f,P32,T32,morc,h32,s32,Ex32,Ex32_fisica,Ex32_quimica)
    c[33]=Corriente(33,f,P33,T33,morc,h33,s33,Ex33,Ex33_fisica,Ex33_quimica)
    c[34]=Corriente(34,f,P34,T34,morc,h34,s34,Ex34,Ex34_fisica,Ex34_quimica)
    c[38]=Corriente(38,"agua",P38,T38,m38,h38,s38,Ex38,Ex38_fisica,Ex38_quimica)
    c[39]=Corriente(39,"agua",P39,T39,m38,h39,s39,Ex39,Ex39_fisica,Ex39_quimica)

    df=pd.DataFrame([{
        "Nodo":c[k].n,
        "Fluido":c[k].f,
        "P(kPa)":c[k].P,
        "T(ºC)":c[k].T,
        "m(kg/s)":c[k].m,
        "h(kJ/kg)":c[k].h,
        "s(kJ/kg K)":c[k].s,
        "Ex(kW)":c[k].Ex,
        "Ex_fisica(kW)":c[k].Ex_fisica,
        "Ex_quimica(kW)":c[k].Ex_quimica
    } for k in [31,32,33,34,38,39]])
    print(df.to_string(index=False))

    p["P33"]=P33  # kPa
    p["P32"]=P32  # kPa
    p["P34"]=P34  # kPa
    p["m_orc"]=morc  # kg/s
    p["Qdisp_orc"]=Qdisp  # kW
    p["Qin_orc"]=Qin  # kW
    p["Qc_orc"]=Qc  # kW
    p["WT_orc"]=WT  # kW
    p["Wb_orc"]=Wb  # kW
    p["Wn_orc"]=Wn  # kW
    p["We_orc"]=We  # kW
    p["m38"]=m38  # kg/s
    p["T39_calc"]=c[39].T  # °C

    return c

corr_orc=ejecutar_orc(parametros_base,corr_efgt)
```

```python
import pandas as pd
from CoolProp.CoolProp import PropsSI
import math

def _w_TP(T,P):
    TK=T+273.15  # K
    PP=P*1000.0  # Pa
    h=PropsSI("H","T",TK,"P",PP,"Water")/1000.0  # kJ/kg
    s=PropsSI("S","T",TK,"P",PP,"Water")/1000.0  # kJ/kgK
    return h,s

def _w_PH(P,h):
    PP=P*1000.0  # Pa
    hh=h*1000.0  # J/kg
    TK=PropsSI("T","P",PP,"H",hh,"Water")
    s=PropsSI("S","P",PP,"H",hh,"Water")/1000.0  # kJ/kgK
    return TK-273.15,s  # °C

def _w_Px(P,x):
    PP=P*1000.0  # Pa
    h=PropsSI("H","P",PP,"Q",x,"Water")/1000.0  # kJ/kg
    s=PropsSI("S","P",PP,"Q",x,"Water")/1000.0  # kJ/kgK
    TK=PropsSI("T","P",PP,"Q",x,"Water")
    return TK-273.15,h,s  # °C

def _Psat_T(T):
    TK=T+273.15  # K
    return PropsSI("P","T",TK,"Q",0.0,"Water")/1000.0  # kPa

def _exfis(m,h,s,h0,s0,T0):
    T0K=T0+273.15  # K
    return m*((h-h0)-T0K*(s-s0))  # kW

def _bpe(t,S):
    A=-4.584e-4*t*t+2.823e-1*t+17.95
    B=1.536e-4*t*t+5.267e-2*t+6.56
    return A*S*S+B*S

def _exquim_sal(m,S,Sref,T0):
    Mw=18.01528  # kg/kmol
    Ms=58.44  # kg/kmol
    R=8.314462618  # kJ/kmolK
    T0K=T0+273.15  # K
    eps=1e-12  # -

    S=max(eps,min(1.0-eps,S))
    Sref=max(eps,min(1.0-eps,Sref))

    nw=(1.0-S)/Mw  # kmol/kg
    ns=S/Ms  # kmol/kg
    nt=nw+ns  # kmol/kg
    xw=nw/nt  # -
    xs=ns/nt  # -

    nw0=(1.0-Sref)/Mw  # kmol/kg
    ns0=Sref/Ms  # kmol/kg
    nt0=nw0+ns0  # kmol/kg
    xw0=nw0/nt0  # -
    xs0=ns0/nt0  # -

    termw=nw*math.log(xw/xw0)  # kmol/kg
    if S<=1e-10:
        terms=0.0  # kmol/kg
    else:
        terms=ns*math.log(xs/xs0)  # kmol/kg

    eq=R*T0K*(termw+terms)  # kJ/kg
    return m*eq  # kW

def ejecutar_med(p,c_bcc):
    T0=p["T0"]  # °C
    P0=p["P0"]  # kPa

    n=int(p.get("n_efectos_med",4))  # -
    if n!=4:
        raise ValueError("n_efectos_med=4")

    P41=float(p.get("P41",2000.0))  # kPa
    P42=float(p.get("P42",P41))  # kPa

    Patm=float(p.get("P_atm",101.3))  # kPa

    T41=float(p.get("T41",53.0))  # °C
    T43=float(p.get("T43",T0))  # °C
    Ttop=float(p.get("T48",51.0))  # °C
    Tbot=float(p.get("T58",39.0))  # °C

    m41=float(p.get("m41",4.043))  # kg/s
    m43=float(p.get("m43",p.get("m44",3.444)))  # kg/s
    m44=float(p.get("m44",m43))  # kg/s
    frac46=float(p.get("frac46",0.25))  # -
    m46=float(p.get("m46",frac46*m44))  # kg/s
    m59=float(p.get("m59",0.4447))  # kg/s

    X43=float(p.get("X43",35000.0))  # ppm
    S43=max(0.0,X43/1e6)  # kg/kg
    Sref=float(p.get("Sref_med",S43))  # kg/kg

    if 6 not in c_bcc or 7 not in c_bcc:
        raise KeyError("faltan_6_7_bcc")
    c6=c_bcc[6]
    c7=c_bcc[7]

    Qin=float(c6.m)*(float(c6.h)-float(c7.h))  # kW
    if Qin<=0:
        raise ValueError("Qin<=0")

    h0,s0=_w_TP(T0,P0)

    h41,s41=_w_TP(T41,P41)
    h42=h41+Qin/m41
    T42,s42=_w_PH(P42,h42)

    Qlazo=m41*(h42-h41)  # kW

    h43,s43=_w_TP(T43,Patm)
    h44=h43+Qlazo/m43
    T44,s44=_w_PH(Patm,h44)

    m45=m44-m46

    Tb=[Ttop-(Ttop-Tbot)*i/(n-1) for i in range(n)]  # °C

    Pe=[0.0]*n
    mb=[]
    dv=[0.0]*n
    Sbr=[0.0]*n
    BPE=[0.0]*n
    Tv=[0.0]*n
    hv=[0.0]*n
    sv=[0.0]*n
    hl=[0.0]*n
    sl=[0.0]*n
    hb=[0.0]*n
    sb=[0.0]*n
    r=[0.0]*n

    mb0=float(m46)
    msal=m46*S43  # kg/s

    for i in range(n):
        mb_tmp=mb0-m59/n
        if mb_tmp<=0:
            mb_tmp=mb0*0.8

        Sbr[i]=min(0.12,max(0.0,msal/mb_tmp))
        BPE[i]=_bpe(Tb[i],Sbr[i])

        Td=Tb[i]-BPE[i]
        Pe[i]=_Psat_T(Td)

        Tvi,hgi,sgi=_w_Px(Pe[i],1.0)
        Tli,hfi,sfi=_w_Px(Pe[i],0.0)
        hbi,sbi=_w_TP(Tb[i],Pe[i])

        Tv[i]=Tvi
        hv[i]=hgi
        sv[i]=sgi
        hl[i]=hfi
        sl[i]=sfi
        hb[i]=hbi
        sb[i]=sbi
        r[i]=max(1e-9,hgi-hfi)

        mb0=mb_tmp

    Qef=[0.0]*n
    Qef[0]=Qlazo

    for i in range(n):
        dv[i]=Qef[i]/r[i]
        if i<n-1:
            Qef[i+1]=dv[i]*r[i]

    escala=m59/sum(dv)
    dv=[di*escala for di in dv]

    mb0=float(m46)
    for i in range(n):
        mb1=mb0-dv[i]
        mb.append(mb1)
        mb0=mb1

    c={}

    Ex41_f=_exfis(m41,h41,s41,h0,s0,T0)
    Ex42_f=_exfis(m41,h42,s42,h0,s0,T0)

    c[41]=Corriente(41,"agua",P41,T41,m41,h41,s41,Ex41_f,Ex41_f,0.0)
    c[42]=Corriente(42,"agua",P42,T42,m41,h42,s42,Ex42_f,Ex42_f,0.0)

    Ex43_f=_exfis(m43,h43,s43,h0,s0,T0)
    Ex43_q=_exquim_sal(m43,S43,Sref,T0)
    c[43]=Corriente(43,"agua_mar",Patm,T43,m43,h43,s43,Ex43_f+Ex43_q,Ex43_f,Ex43_q)

    Ex44_f=_exfis(m44,h44,s44,h0,s0,T0)
    Ex44_q=_exquim_sal(m44,S43,Sref,T0)
    c[44]=Corriente(44,"agua_mar",Patm,T44,m44,h44,s44,Ex44_f+Ex44_q,Ex44_f,Ex44_q)

    Ex45_f=_exfis(m45,h44,s44,h0,s0,T0)
    Ex45_q=_exquim_sal(m45,S43,Sref,T0)
    c[45]=Corriente(45,"agua_mar",Patm,T44,m45,h44,s44,Ex45_f+Ex45_q,Ex45_f,Ex45_q)

    Ex46_f=_exfis(m46,h44,s44,h0,s0,T0)
    Ex46_q=_exquim_sal(m46,S43,Sref,T0)
    c[46]=Corriente(46,"agua_mar",Patm,T44,m46,h44,s44,Ex46_f+Ex46_q,Ex46_f,Ex46_q)

    n_v=[47,50,53,56]
    n_d=[49,52,55,57]
    n_b=[48,51,54,58]

    for i in range(n):
        mv=dv[i]
        md=dv[i]
        mbri=mb[i]

        Exv_f=_exfis(mv,hv[i],sv[i],h0,s0,T0)
        Exv_q=_exquim_sal(mv,0.0,Sref,T0)

        Exd_f=_exfis(md,hl[i],sl[i],h0,s0,T0)
        Exd_q=_exquim_sal(md,0.0,Sref,T0)

        Exb_f=_exfis(mbri,hb[i],sb[i],h0,s0,T0)
        Exb_q=_exquim_sal(mbri,Sbr[i],Sref,T0)

        c[n_v[i]]=Corriente(n_v[i],"vapor",Pe[i],Tv[i],mv,hv[i],sv[i],Exv_f+Exv_q,Exv_f,Exv_q)
        c[n_d[i]]=Corriente(n_d[i],"destilado",Pe[i],Tv[i],md,hl[i],sl[i],Exd_f+Exd_q,Exd_f,Exd_q)
        c[n_b[i]]=Corriente(n_b[i],"salmuera",Pe[i],Tb[i],mbri,hb[i],sb[i],Exb_f+Exb_q,Exb_f,Exb_q)

    h59=(c[49].m*c[49].h+c[52].m*c[52].h+c[55].m*c[55].h+c[57].m*c[57].h)/m59
    T59,s59=_w_PH(Patm,h59)
    Ex59_f=_exfis(m59,h59,s59,h0,s0,T0)
    Ex59_q=_exquim_sal(m59,0.0,Sref,T0)

    c[59]=Corriente(59,"destilado",Patm,T59,m59,h59,s59,Ex59_f+Ex59_q,Ex59_f,Ex59_q)

    df=pd.DataFrame([{
        "Nodo":ci.n,
        "Fluido":ci.f,
        "P(kPa)":ci.P,
        "T(°C)":ci.T,
        "m(kg/s)":ci.m,
        "h(kJ/kg)":ci.h,
        "s(kJ/kg K)":ci.s,
        "Ex(kW)":ci.Ex,
        "Ex_fisica(kW)":ci.Ex_fisica,
        "Ex_quimica(kW)":ci.Ex_quimica
    } for ci in c.values()])
    print(df.to_string(index=False))

    return c

corr_med=ejecutar_med(parametros_base,corr_bcc)
```

```python
import pandas as pd
from CoolProp.CoolProp import PropsSI
import numpy as np

class LiBr:
    A=np.array([
        [-9.133128E0,-4.759724E-1,-5.638171E-2,1.108418E-3],
        [ 9.439697E-1,-2.882015E-3,-1.345435E-4,5.852133E-7],
        [-7.324352E-5,-1.556533E-5, 1.992657E-6,-3.924205E-8],
    ],dtype=float)
    B=np.array([
        [ 5.127558E-01,-1.393954E-02, 2.924145E-05, 9.035697E-07],
        [ 1.226780E-02,-9.156820E-05, 1.820453E-08,-7.991806E-10],
        [-1.364895E-05, 1.068904E-07,-1.381109E-09, 1.529784E-11],
        [ 1.021501E-08, 0.0,          0.0,          0.0],
    ],dtype=float)
    k0=7.05
    k1=-1603.54
    k2=-104095.5
    A0=3.462023
    A1=-2.679895e-2
    B0=1.3499e-3
    B1=-6.55e-6
    D0=162.81
    D1=-6.0418
    D2=-4.5348e-3
    D3=1.2053e-3

    @staticmethod
    def TD(T,Xw):
        X=float(Xw)
        T=float(T)
        td=0.0
        for i in range(3):
            for j in range(4):
                td+=float(LiBr.A[i,j])*((X-40.0)**j)*(T**i)
        return float(td)

    @staticmethod
    def Pv(T,Xw):
        td=LiBr.TD(T,Xw)
        logP=float(LiBr.k0)+float(LiBr.k1)/(td+273.15)+float(LiBr.k2)/(td+273.15)**2
        return float(10.0**logP)

    @staticmethod
    def h(T,Xw):
        T=float(T)
        X=float(Xw)
        return float((LiBr.A0+LiBr.A1*X)*T+0.5*(LiBr.B0+LiBr.B1*X)*T**2+(LiBr.D0+LiBr.D1*X+LiBr.D2*X**2+LiBr.D3*X**3))

    @staticmethod
    def s(T,Xw):
        T=float(T)
        X=float(Xw)
        ss=0.0
        for i in range(4):
            for j in range(4):
                ss+=float(LiBr.B[i,j])*(X**j)*(T**i)
        return float(ss)

    @staticmethod
    def xPT(T,P):
        T=float(T)
        P=float(P)
        if P<=0.0:
            return []
        logP=float(np.log10(P))
        a=float(LiBr.k0-logP)
        b=float(LiBr.k1)
        c=float(LiBr.k2)
        eps=1e-14
        TDs=[]
        if abs(a)<eps:
            if abs(b)<eps:
                return []
            TDs=[-c/b]
        else:
            disc=b*b-4.0*a*c
            if disc<0.0:
                return []
            sd=float(np.sqrt(disc))
            TDs=[(-b+sd)/(2.0*a),(-b-sd)/(2.0*a)]
        sols=[]
        A=LiBr.A
        for TDk in TDs:
            TD=float(TDk-273.15)
            c3=float(A[0,3]+A[1,3]*T+A[2,3]*(T**2))
            c2=float(A[0,2]+A[1,2]*T+A[2,2]*(T**2))
            c1=float(A[0,1]+A[1,1]*T+A[2,1]*(T**2))
            c0=float(A[0,0]+A[1,0]*T+A[2,0]*(T**2)-TD)
            r=np.roots([c3,c2,c1,c0])
            for rr in r:
                if abs(np.imag(rr))<=1e-9:
                    Xw=float(np.real(rr)+40.0)
                    if 40.0<=Xw<=70.0:
                        sols.append(Xw/100.0)
        u=[]
        for xf in sols:
            if not any(abs(xf-uu)<1e-9 for uu in u):
                u.append(float(xf))
        return u

def _wTP(T,P):
    TK=T+273.15  # K
    PP=P*1000.0  # Pa
    h=PropsSI("H","T",TK,"P",PP,"Water")/1000.0  # kJ/kg
    s=PropsSI("S","T",TK,"P",PP,"Water")/1000.0  # kJ/kgK
    return float(h),float(s)

def _wTQ(T,Q):
    TK=T+273.15  # K
    h=PropsSI("H","T",TK,"Q",float(Q),"Water")/1000.0  # kJ/kg
    s=PropsSI("S","T",TK,"Q",float(Q),"Water")/1000.0  # kJ/kgK
    P=PropsSI("P","T",TK,"Q",float(Q),"Water")/1000.0  # kPa
    return float(P),float(h),float(s)

def _wPH(P,h):
    PP=P*1000.0  # Pa
    hJ=h*1000.0  # J/kg
    T=PropsSI("T","P",PP,"H",hJ,"Water")-273.15  # °C
    s=PropsSI("S","P",PP,"H",hJ,"Water")/1000.0  # kJ/kgK
    return float(T),float(s)

def _ex(m,h,s,h0,s0,T0):
    T0K=T0+273.15  # K
    ex=(h-h0)-T0K*(s-s0)  # kJ/kg
    return float(m)*float(ex)  # kW

def _T_Pv_X(PkPa,x,Tlo=0.0,Thi=180.0):
    a=float(Tlo)
    b=float(Thi)
    fa=LiBr.Pv(a,100.0*x)-PkPa
    fb=LiBr.Pv(b,100.0*x)-PkPa
    if fa==0.0:
        return float(a)
    if fb==0.0:
        return float(b)
    if fa*fb>0.0:
        return float(a if abs(fa)<abs(fb) else b)
    for _ in range(120):
        m=0.5*(a+b)
        fm=LiBr.Pv(m,100.0*x)-PkPa
        if fa*fm<=0.0:
            b=m
            fb=fm
        else:
            a=m
            fa=fm
    return float(0.5*(a+b))

def _xsxw(mr,T61,Pb,T64,Pa,xs_obj=None,xw_obj=None):
    xw=LiBr.xPT(T61,Pb)
    xs=LiBr.xPT(T64,Pa)
    if (not xw) or (not xs):
        raise ValueError("xPT_sin_solucion")
    xw=[x for x in xw if 0.40<=x<=0.60] or xw
    xs=[x for x in xs if 0.50<=x<=0.70] or xs
    mejor=None
    mejor_score=float("inf")
    for xss in xs:
        for xww in xw:
            if xss<=xww+1e-12:
                continue
            mw=mr*xss/(xss-xww)
            ms=mw-mr
            if mw<=0.0 or ms<=0.0:
                continue
            score=0.0
            if xs_obj is not None:
                score+=100.0*abs(xss-float(xs_obj))
            else:
                score+=10.0*abs(xss-0.62)
            if xw_obj is not None:
                score+=100.0*abs(xww-float(xw_obj))
            else:
                score+=10.0*abs(xww-0.563)
            score+=1e-3*mw
            if score<mejor_score:
                mejor_score=score
                mejor=(float(ms),float(mw),float(xss),float(xww))
    if mejor is None:
        raise ValueError("no_par_xs_xw")
    return mejor

def _Wb(m,P1,P2,eta,rho):
    if eta<=0.0:
        raise ValueError("eta_bomba_invalida")
    dP=(P2-P1)*1000.0  # Pa
    v=1.0/rho  # m3/kg
    wideal=(v*dP)/1000.0  # kJ/kg
    return float(m)*float(wideal)/float(eta)  # kW

def ejecutar_seac(p):
    T0=p["T0"]  # °C
    P0=p["P0"]  # kPa

    Qev=p.get("Q_ev_seac",p.get("Q_evaporador_seac",77.43))  # kW
    etahx=p.get("eta_hx_seac",p.get("eta_intercambiador_seac",0.68))  # -
    etab=p.get("eta_b_seac",p.get("eta_bomba_seac",0.70))  # -
    dTpp=p.get("dTpp_gen",p.get("deltaT_pp_generador_seac",10.0))  # °C
    rho=p.get("rho_libr",1200.0)  # kg/m3

    T61=p["T61"]  # °C
    T62=p["T62"]  # °C
    T64=p["T64"]  # °C

    T68=p["T68"]  # °C
    T69=p["T69"]  # °C
    T70=p["T70"]  # °C
    T71=p["T71"]  # °C
    T72=p["T72"]  # °C
    T73=p["T73"]  # °C
    T74=p["T74"]  # °C
    T75=p["T75"]  # °C
    T76=p["T76"]  # °C

    T67=float(p.get("T67",T68+dTpp))  # °C

    P71=p.get("P71",P0)  # kPa
    P72=p.get("P72",P71)  # kPa
    P73=p.get("P73",P0)  # kPa
    P74=p.get("P74",100.0)  # kPa
    P75=p.get("P75",150.0)  # kPa
    P76=p.get("P76",P75)  # kPa

    Pb,_,_=_wTQ(T69,0.0)
    Pa,h68,s68=_wTQ(T68,0.0)
    P67=Pa  # kPa

    h67,s67=_wTP(T67,P67)

    usar_valvula_isoentalpica=bool(p.get("usar_valvula_isoentalpica",True))  # -
    if usar_valvula_isoentalpica:
        h69=h68
        T69_calc,s69=_wPH(Pb,h69)
        T69=float(T69_calc)
    else:
        _,h69,s69=_wTQ(T69,0.0)

    _,h70,s70=_wTQ(T70,1.0)

    den=h70-h69  # kJ/kg
    if den<=0.0:
        raise ValueError("h70_h69_invalido")
    mr=Qev/den  # kg/s
    if mr<=0.0:
        raise ValueError("mr<=0")

    xs_obj=p.get("xs_obj",None)  # -
    xw_obj=p.get("xw_obj",None)  # -
    ms,mw,xs,xw=_xsxw(mr,T61,Pb,T64,Pa,xs_obj=xs_obj,xw_obj=xw_obj)

    Xw=100.0*xw  # %
    Xs=100.0*xs  # %

    if etahx<0.0 or etahx>1.0:
        raise ValueError("eta_hx_invalida")

    T65=float(p.get("T65",etahx*T62+(1.0-etahx)*T64))  # °C

    h61=LiBr.h(T61,Xw)  # kJ/kg
    s61=LiBr.s(T61,Xw)  # kJ/kgK
    h62=LiBr.h(T62,Xw)  # kJ/kg
    s62=LiBr.s(T62,Xw)  # kJ/kgK

    h64=LiBr.h(T64,Xs)  # kJ/kg
    s64=LiBr.s(T64,Xs)  # kJ/kgK
    h65=LiBr.h(T65,Xs)  # kJ/kg
    s65=LiBr.s(T65,Xs)  # kJ/kgK

    h63=h62+etahx*(ms/mw)*(h64-h65)  # kJ/kg

    a=0.0
    b=180.0
    fa=LiBr.h(a,Xw)-h63
    fb=LiBr.h(b,Xw)-h63
    if fa*fb>0.0:
        T63=float(a if abs(fa)<abs(fb) else b)
    else:
        for _ in range(120):
            m=0.5*(a+b)
            fm=LiBr.h(m,Xw)-h63
            if fa*fm<=0.0:
                b=m
                fb=fm
            else:
                a=m
                fa=fm
        T63=float(0.5*(a+b))
    s63=LiBr.s(T63,Xw)  # kJ/kgK

    usar_T66_P=bool(p.get("usar_T66_por_P",True))  # -
    if usar_T66_P:
        T66=_T_Pv_X(Pb,xs,0.0,180.0)
        h66=h65
        s66=LiBr.s(T66,Xs)  # kJ/kgK
    else:
        T66=T65
        h66=h65
        s66=s65

    Wb=_Wb(mw,Pb,Pa,etab,rho)  # kW

    Qc=mr*(h67-h68)  # kW
    Qab=ms*h66+mr*h70-mw*h61  # kW
    Qg=mr*h67+ms*h64-mw*h63  # kW
    COP=Qev/Qg if abs(Qg)>1e-12 else float("nan")  # -

    h0w,s0w=_wTP(T0,P0)
    h0w_libr=LiBr.h(T0,Xw)  # kJ/kg
    s0w_libr=LiBr.s(T0,Xw)  # kJ/kgK
    h0s_libr=LiBr.h(T0,Xs)  # kJ/kg
    s0s_libr=LiBr.s(T0,Xs)  # kJ/kgK

    Ex61_fisica=_ex(mw,h61,s61,h0w_libr,s0w_libr,T0)  # kW
    Ex62_fisica=_ex(mw,h62,s62,h0w_libr,s0w_libr,T0)  # kW
    Ex63_fisica=_ex(mw,h63,s63,h0w_libr,s0w_libr,T0)  # kW

    Ex64_fisica=_ex(ms,h64,s64,h0s_libr,s0s_libr,T0)  # kW
    Ex65_fisica=_ex(ms,h65,s65,h0s_libr,s0s_libr,T0)  # kW
    Ex66_fisica=_ex(ms,h66,s66,h0s_libr,s0s_libr,T0)  # kW

    Ex67_fisica=_ex(mr,h67,s67,h0w,s0w,T0)  # kW
    Ex68_fisica=_ex(mr,h68,s68,h0w,s0w,T0)  # kW
    Ex69_fisica=_ex(mr,h69,s69,h0w,s0w,T0)  # kW
    Ex70_fisica=_ex(mr,h70,s70,h0w,s0w,T0)  # kW

    h71,s71=_wTP(T71,P71)
    h72,s72=_wTP(T72,P72)
    h73,s73=_wTP(T73,P73)
    h74,s74=_wTP(T74,P74)

    d1=h72-h71  # kJ/kg
    d2=h74-h73  # kJ/kg
    if abs(d1)<1e-12 or abs(d2)<1e-12:
        raise ValueError("delta_h_lazos_invalido")

    m71=Qc/d1  # kg/s
    m73=Qab/d2  # kg/s
    if m71<=0.0 or m73<=0.0:
        raise ValueError("m71_o_m73<=0")

    Ex71_fisica=_ex(m71,h71,s71,h0w,s0w,T0)  # kW
    Ex72_fisica=_ex(m71,h72,s72,h0w,s0w,T0)  # kW
    Ex73_fisica=_ex(m73,h73,s73,h0w,s0w,T0)  # kW
    Ex74_fisica=_ex(m73,h74,s74,h0w,s0w,T0)  # kW

    h75,s75=_wTP(T75,P75)
    h76,s76=_wTP(T76,P76)
    Qcal=p.get("Qcal",p.get("Q_calefaccion",p.get("Q_calefaccion_obj",172.4)))  # kW
    d3=h75-h76  # kJ/kg
    if abs(d3)<1e-12:
        raise ValueError("h75_h76_invalido")
    m75=Qcal/d3  # kg/s
    if m75<=0.0:
        raise ValueError("m75<=0")

    Ex75_fisica=_ex(m75,h75,s75,h0w,s0w,T0)  # kW
    Ex76_fisica=_ex(m75,h76,s76,h0w,s0w,T0)  # kW

    Ex61_quimica=0.0  # kW
    Ex62_quimica=0.0  # kW
    Ex63_quimica=0.0  # kW
    Ex64_quimica=0.0  # kW
    Ex65_quimica=0.0  # kW
    Ex66_quimica=0.0  # kW
    Ex67_quimica=0.0  # kW
    Ex68_quimica=0.0  # kW
    Ex69_quimica=0.0  # kW
    Ex70_quimica=0.0  # kW
    Ex71_quimica=0.0  # kW
    Ex72_quimica=0.0  # kW
    Ex73_quimica=0.0  # kW
    Ex74_quimica=0.0  # kW
    Ex75_quimica=0.0  # kW
    Ex76_quimica=0.0  # kW

    c={}
    c[61]=Corriente(61,"LiBr",Pb,T61,mw,h61,s61,Ex61_fisica+Ex61_quimica,Ex61_fisica,Ex61_quimica)
    c[62]=Corriente(62,"LiBr",Pa,T62,mw,h62,s62,Ex62_fisica+Ex62_quimica,Ex62_fisica,Ex62_quimica)
    c[63]=Corriente(63,"LiBr",Pa,T63,mw,h63,s63,Ex63_fisica+Ex63_quimica,Ex63_fisica,Ex63_quimica)

    c[64]=Corriente(64,"LiBr",Pa,T64,ms,h64,s64,Ex64_fisica+Ex64_quimica,Ex64_fisica,Ex64_quimica)
    c[65]=Corriente(65,"LiBr",Pa,T65,ms,h65,s65,Ex65_fisica+Ex65_quimica,Ex65_fisica,Ex65_quimica)
    c[66]=Corriente(66,"LiBr",Pb,T66,ms,h66,s66,Ex66_fisica+Ex66_quimica,Ex66_fisica,Ex66_quimica)

    c[67]=Corriente(67,"agua",P67,T67,mr,h67,s67,Ex67_fisica+Ex67_quimica,Ex67_fisica,Ex67_quimica)
    c[68]=Corriente(68,"agua",Pa,T68,mr,h68,s68,Ex68_fisica+Ex68_quimica,Ex68_fisica,Ex68_quimica)
    c[69]=Corriente(69,"agua",Pb,T69,mr,h69,s69,Ex69_fisica+Ex69_quimica,Ex69_fisica,Ex69_quimica)
    c[70]=Corriente(70,"agua",Pb,T70,mr,h70,s70,Ex70_fisica+Ex70_quimica,Ex70_fisica,Ex70_quimica)

    c[71]=Corriente(71,"agua",P71,T71,m71,h71,s71,Ex71_fisica+Ex71_quimica,Ex71_fisica,Ex71_quimica)
    c[72]=Corriente(72,"agua",P72,T72,m71,h72,s72,Ex72_fisica+Ex72_quimica,Ex72_fisica,Ex72_quimica)
    c[73]=Corriente(73,"agua",P73,T73,m73,h73,s73,Ex73_fisica+Ex73_quimica,Ex73_fisica,Ex73_quimica)
    c[74]=Corriente(74,"agua",P74,T74,m73,h74,s74,Ex74_fisica+Ex74_quimica,Ex74_fisica,Ex74_quimica)

    c[75]=Corriente(75,"agua",P75,T75,m75,h75,s75,Ex75_fisica+Ex75_quimica,Ex75_fisica,Ex75_quimica)
    c[76]=Corriente(76,"agua",P76,T76,m75,h76,s76,Ex76_fisica+Ex76_quimica,Ex76_fisica,Ex76_quimica)

    df=pd.DataFrame([{
        "Nodo":c[k].n,
        "Fluido":c[k].f,
        "P(kPa)":c[k].P,
        "T(ºC)":c[k].T,
        "m(kg/s)":c[k].m,
        "h(kJ/kg)":c[k].h,
        "s(kJ/kg K)":c[k].s,
        "Ex(kW)":c[k].Ex,
        "Ex_fisica(kW)":c[k].Ex_fisica,
        "Ex_quimica(kW)":c[k].Ex_quimica
    } for k in [61,62,63,64,65,66,67,68,69,70,71,72,73,74,75,76]])
    print(df.to_string(index=False))

    p["Pb_seac"]=Pb  # kPa
    p["Pa_seac"]=Pa  # kPa
    p["mr"]=mr  # kg/s
    p["ms"]=ms  # kg/s
    p["mw"]=mw  # kg/s
    p["xs"]=xs  # -
    p["xw"]=xw  # -
    p["T63"]=T63  # °C
    p["T65"]=T65  # °C
    p["T66"]=T66  # °C
    p["T67"]=T67  # °C
    p["Qc"]=Qc  # kW
    p["Qab"]=Qab  # kW
    p["Qg"]=Qg  # kW
    p["Wb"]=Wb  # kW
    p["COP"]=COP  # -

    return c

corr_seac=ejecutar_seac(parametros_base)
```

```python
import pandas as pd

def _mh(c):
    return float(c.m)*float(c.h)  # kW

def _Q_1in1out(c_in,c_out):
    return _mh(c_out)-_mh(c_in)  # kW

def _W_comp_mag(c_in,c_out):
    return _mh(c_out)-_mh(c_in)  # kW

def _W_turb_mag(c_in,c_out):
    return _mh(c_in)-_mh(c_out)  # kW

def _Q_gen_1in2out(c_in,c_out1,c_out2):
    return _mh(c_out1)+_mh(c_out2)-_mh(c_in)  # kW

def _Q_abs_2in1out(c_in1,c_in2,c_out):
    return _mh(c_out)-(_mh(c_in1)+_mh(c_in2))  # kW

def _residuo(entradas,salidas,Q,W_signed,c):
    mh_in=sum(_mh(c[i]) for i in entradas)  # kW
    mh_out=sum(_mh(c[i]) for i in salidas)  # kW
    dmh=mh_out-mh_in  # kW
    return float(dmh-float(Q)+float(W_signed))  # kW

def calcular_QW_subunidades(p,c):
    r=[]

    def add(nom,tipo,entradas,salidas,Q,W_signed):
        res=_residuo(entradas,salidas,Q,W_signed,c)  # kW
        r.append({
            "Subunidad":str(nom),
            "Tipo":str(tipo),
            "Q(kW)":float(Q),
            "W(kW)":float(W_signed),
            "Residuo(kW)":float(res),
        })

    if 5 in c and 6 in c:
        Q=_Q_1in1out(c[5],c[6])  # kW
        add("BCC_HTHE(5->6)","HX",entradas=[5],salidas=[6],Q=Q,W_signed=0.0)

    if 6 in c and 7 in c:
        Q=_Q_1in1out(c[6],c[7])  # kW
        add("Calentador_MED(6->7)","HX",entradas=[6],salidas=[7],Q=Q,W_signed=0.0)

    if 11 in c and 12 in c:
        W=-_W_comp_mag(c[11],c[12])  # kW
        add("Compresor(11->12)","COMP",entradas=[11],salidas=[12],Q=0.0,W_signed=W)

    if 12 in c and 13 in c:
        Q=_Q_1in1out(c[12],c[13])  # kW
        add("Precalentador_solar(12->13)","HX",entradas=[12],salidas=[13],Q=Q,W_signed=0.0)

    if 13 in c and 14 in c:
        Q=_Q_1in1out(c[13],c[14])  # kW
        add("HTHE_altaT(13->14)","HX",entradas=[13],salidas=[14],Q=Q,W_signed=0.0)

    if 14 in c and 15 in c:
        W=_W_turb_mag(c[14],c[15])  # kW
        add("Turbina_gas(14->15)","TURB",entradas=[14],salidas=[15],Q=0.0,W_signed=W)

    if 15 in c and 16 in c:
        Q=_Q_1in1out(c[15],c[16])  # kW
        add("Generador_SRC(15->16)","HX",entradas=[15],salidas=[16],Q=Q,W_signed=0.0)

    if 16 in c and 17 in c:
        Q=_Q_1in1out(c[16],c[17])  # kW
        add("Generador_ORC(16->17)","HX",entradas=[16],salidas=[17],Q=Q,W_signed=0.0)

    if 21 in c and 22 in c:
        W=_W_turb_mag(c[21],c[22])  # kW
        add("Turbina_SRC(21->22)","TURB",entradas=[21],salidas=[22],Q=0.0,W_signed=W)

    if 22 in c and 23 in c:
        Q=_Q_1in1out(c[22],c[23])  # kW
        add("Condensador_SRC(22->23)","HX",entradas=[22],salidas=[23],Q=Q,W_signed=0.0)

    if 23 in c and 24 in c:
        W=-_W_comp_mag(c[23],c[24])  # kW
        add("Bomba_SRC(23->24)","PUMP",entradas=[23],salidas=[24],Q=0.0,W_signed=W)

    if 28 in c and 29 in c:
        Q=_Q_1in1out(c[28],c[29])  # kW
        add("DHU(28->29)","HX",entradas=[28],salidas=[29],Q=Q,W_signed=0.0)

    if 34 in c and 31 in c:
        Q=_Q_1in1out(c[34],c[31])  # kW
        add("Evap_ORC(34->31)","HX",entradas=[34],salidas=[31],Q=Q,W_signed=0.0)

    if 31 in c and 32 in c:
        W=_W_turb_mag(c[31],c[32])  # kW
        add("Turbina_ORC(31->32)","TURB",entradas=[31],salidas=[32],Q=0.0,W_signed=W)

    if 32 in c and 33 in c:
        Q=_Q_1in1out(c[32],c[33])  # kW
        add("Cond_ORC(32->33)","HX",entradas=[32],salidas=[33],Q=Q,W_signed=0.0)

    if 33 in c and 34 in c:
        W=-_W_comp_mag(c[33],c[34])  # kW
        add("Bomba_ORC(33->34)","PUMP",entradas=[33],salidas=[34],Q=0.0,W_signed=W)

    if 38 in c and 39 in c:
        Q=_Q_1in1out(c[38],c[39])  # kW
        add("Lazo_agua_ORC(38->39)","HX",entradas=[38],salidas=[39],Q=Q,W_signed=0.0)

    if 41 in c and 42 in c:
        Q=_Q_1in1out(c[41],c[42])  # kW
        add("Calentador_agua_MED(41->42)","HX",entradas=[41],salidas=[42],Q=Q,W_signed=0.0)

    if 69 in c and 70 in c:
        Q=_Q_1in1out(c[69],c[70])  # kW
        add("Evap_SEAC(69->70)","HX",entradas=[69],salidas=[70],Q=Q,W_signed=0.0)

    if 67 in c and 68 in c:
        Q=_Q_1in1out(c[67],c[68])  # kW
        add("Cond_SEAC(67->68)","HX",entradas=[67],salidas=[68],Q=Q,W_signed=0.0)

    if 66 in c and 70 in c and 61 in c:
        Q=_Q_abs_2in1out(c[66],c[70],c[61])  # kW
        add("Abs_SEAC(66,70->61)","MIX",entradas=[66,70],salidas=[61],Q=Q,W_signed=0.0)

    if 64 in c and 63 in c and 67 in c:
        Q=_Q_gen_1in2out(c[64],c[63],c[67])  # kW
        add("Gen_SEAC(64->63,67)","SPLIT",entradas=[64],salidas=[63,67],Q=Q,W_signed=0.0)

    if 61 in c and 62 in c:
        W=-_W_comp_mag(c[61],c[62])  # kW
        add("Bomba_SEAC(61->62)","PUMP",entradas=[61],salidas=[62],Q=0.0,W_signed=W)

    df=pd.DataFrame(r)
    print(df.to_string(index=False))
    return df

c_total={**corr_bcc,**corr_efgt,**corr_src,**corr_orc,**corr_med,**corr_seac,**corr_dhu}
df_QW=calcular_QW_subunidades(parametros_base,c_total)
```

```python
import pandas as pd

def _getW(df_QW,nombre):
    v=df_QW.loc[df_QW["Subunidad"]==nombre,"W(kW)"]
    return float(v.iloc[0]) if len(v) else 0.0

def _getQ(df_QW,nombre):
    v=df_QW.loc[df_QW["Subunidad"]==nombre,"Q(kW)"]
    return float(v.iloc[0]) if len(v) else 0.0

def _get_param(parametros,keys,default=0.0):
    for k in keys:
        if k in parametros:
            try:
                return float(parametros[k])
            except Exception:
                pass
    return float(default)

def calcular_eficiencias_energeticas(parametros,df_QW):
    Qsolar_in=_get_param(parametros,["Qsolar_util","Qsolar_util_kW"],0.0)  # kW

    m3=_get_param(parametros,["m3"],0.0)  # kg/s
    mgn=_get_param(parametros,["mgn"],0.0)  # kg/s
    pci_b=_get_param(parametros,["PCI_bagazo","PCI_bagazo_kJ_kg"],0.0)  # kJ/kg
    pci_g=_get_param(parametros,["PCI_gn","PCI_gn_kJ_kg"],0.0)  # kJ/kg

    Qfuel_in=float(m3*pci_b+mgn*pci_g)  # kW
    Qin_total=float(Qsolar_in+Qfuel_in)  # kW

    eta_gen_efgt=_get_param(parametros,["eta_generador_efgt"],1.0)  # -
    eta_gen_src=_get_param(parametros,["eta_generador_src"],1.0)  # -
    eta_gen_orc=_get_param(parametros,["eta_generador_orc"],1.0)  # -

    W_comp=_getW(df_QW,"Compresor(11->12)")  # kW
    W_tg=_getW(df_QW,"Turbina_gas(14->15)")  # kW
    W_efgt_mec=float(W_tg+W_comp)  # kW
    W_efgt_e=float(eta_gen_efgt*W_efgt_mec)  # kW

    W_ts=_getW(df_QW,"Turbina_SRC(21->22)")  # kW
    W_ps=_getW(df_QW,"Bomba_SRC(23->24)")  # kW
    W_src_mec=float(W_ts+W_ps)  # kW
    W_src_e=float(eta_gen_src*W_src_mec)  # kW

    W_to=_getW(df_QW,"Turbina_ORC(31->32)")  # kW
    W_po=_getW(df_QW,"Bomba_ORC(33->34)")  # kW
    W_orc_mec=float(W_to+W_po)  # kW
    W_orc_e=float(eta_gen_orc*W_orc_mec)  # kW

    W_bp=_getW(df_QW,"Bomba_SEAC(61->62)")  # kW
    W_aux=float(W_bp)  # kW

    W_neta=float(W_efgt_e+W_src_e+W_orc_e+W_aux)  # kW

    Q_dhu=float(_getQ(df_QW,"DHU(28->29)"))  # kW
    Q_ev_seac=float(_getQ(df_QW,"Evap_SEAC(69->70)"))  # kW

    m_dest=_get_param(parametros,["m59","m_agua_potable_obj_kgs","m_agua_potable_obj_kg_s","m_agua_potable_kgs"],0.0)  # kg/s
    m3h_dest=float(m_dest*3600.0/1000.0)  # m3/h

    eta_el=float("nan") if Qin_total<=1e-12 else float(W_neta/Qin_total)  # -
    eta_cal=float("nan") if Qin_total<=1e-12 else float(Q_dhu/Qin_total)  # -
    eta_frio=float("nan") if Qin_total<=1e-12 else float(Q_ev_seac/Qin_total)  # -
    eta_energia_util=float("nan") if Qin_total<=1e-12 else float((W_neta+Q_dhu+Q_ev_seac)/Qin_total)  # -

    resumen=pd.DataFrame([
        {"Indicador":"Qsolar_in(kW)","Valor":Qsolar_in},
        {"Indicador":"Qfuel_in(kW)","Valor":Qfuel_in},
        {"Indicador":"Qin_total(kW)","Valor":Qin_total},
        {"Indicador":"W_efgt_mec(kW)","Valor":W_efgt_mec},
        {"Indicador":"W_efgt_e(kW)","Valor":W_efgt_e},
        {"Indicador":"W_src_mec(kW)","Valor":W_src_mec},
        {"Indicador":"W_src_e(kW)","Valor":W_src_e},
        {"Indicador":"W_orc_mec(kW)","Valor":W_orc_mec},
        {"Indicador":"W_orc_e(kW)","Valor":W_orc_e},
        {"Indicador":"W_aux_total(kW)","Valor":W_aux},
        {"Indicador":"W_neta_total(kW)","Valor":W_neta},
        {"Indicador":"Q_DHU(kW)","Valor":Q_dhu},
        {"Indicador":"Q_ev_SEAC(kW)","Valor":Q_ev_seac},
        {"Indicador":"m_destilado(kg/s)","Valor":m_dest},
        {"Indicador":"m_destilado(m3/h)","Valor":m3h_dest},
        {"Indicador":"eta_el(-)","Valor":eta_el},
        {"Indicador":"eta_calefaccion(-)","Valor":eta_cal},
        {"Indicador":"eta_enfriamiento(-)","Valor":eta_frio},
        {"Indicador":"eta_total_util(-)","Valor":eta_energia_util},
    ])

    ficha=pd.DataFrame([{
        "W_neta(kW)":W_neta,
        "Q_DHU(kW)":Q_dhu,
        "Q_ev_SEAC(kW)":Q_ev_seac,
        "m_destilado(kg/s)":m_dest,
        "m_destilado(m3/h)":m3h_dest,
        "Qin_total(kW)":Qin_total
    }])

    print(resumen.to_string(index=False))
    print(ficha.to_string(index=False))
    return resumen,ficha

df_etas,ficha_productos=calcular_eficiencias_energeticas(parametros_base,df_QW)
```

Análisis exergético

```python
import numpy as np
import pandas as pd

def _getW(df_QW,nombre):
    v=df_QW.loc[df_QW["Subunidad"]==nombre,"W(kW)"]
    return float(v.iloc[0]) if len(v) else 0.0

def _getQ(df_QW,nombre):
    v=df_QW.loc[df_QW["Subunidad"]==nombre,"Q(kW)"]
    return float(v.iloc[0]) if len(v) else 0.0

def _attr_o_clave(x,nombres,default=None):
    for nombre in nombres:
        if hasattr(x,nombre):
            try:
                return float(getattr(x,nombre))
            except:
                pass
        if isinstance(x,dict) and nombre in x:
            try:
                return float(x[nombre])
            except:
                pass
    return default

def _Ex(c,n):
    x=c[n]
    v=_attr_o_clave(x,["Ex"],None)
    if v is not None:
        return float(v)  # kW
    Ex_fisica=_attr_o_clave(x,["Ex_fisica","ex_fisica","Ex_ph","ex_ph"],0.0)  # kW
    Ex_quimica=_attr_o_clave(x,["Ex_quimica","ex_quimica","Ex_ch","ex_ch"],0.0)  # kW
    return float(Ex_fisica+Ex_quimica)  # kW

def _Ex_ph(c,n):
    x=c[n]
    v=_attr_o_clave(x,["Ex_fisica","ex_fisica","Ex_ph","ex_ph"],None)
    if v is not None:
        return float(v)  # kW
    v=_attr_o_clave(x,["Ex"],None)
    if v is not None:
        return float(v)  # kW
    return 0.0  # kW

def _Ex_ch(c,n):
    x=c[n]
    v=_attr_o_clave(x,["Ex_quimica","ex_quimica","Ex_ch","ex_ch"],None)
    if v is not None:
        return float(v)  # kW
    return 0.0  # kW

def _Ex_total(c,n):
    x=c[n]
    v_total=_attr_o_clave(x,["Ex"],None)
    v_fisica=_attr_o_clave(x,["Ex_fisica","ex_fisica","Ex_ph","ex_ph"],None)
    v_quimica=_attr_o_clave(x,["Ex_quimica","ex_quimica","Ex_ch","ex_ch"],None)
    if (v_fisica is not None) or (v_quimica is not None):
        return float((0.0 if v_fisica is None else v_fisica)+(0.0 if v_quimica is None else v_quimica))  # kW
    if v_total is not None:
        return float(v_total)  # kW
    return 0.0  # kW

def _Ex_heat(Q_kW,T0_C,Tb_C):
    T0_K=float(T0_C)+273.15  # K
    Tb_K=float(Tb_C)+273.15  # K
    if abs(Q_kW)<1e-12:
        return 0.0  # kW
    if Tb_K<=1e-9:
        return 0.0  # kW
    return float(Q_kW)*(1.0-T0_K/Tb_K)  # kW

def _row(nombre,ExD_kW,Ex_in_kW,Ex_out_kW,ExQ_kW,ExW_kW):
    return {
        "Equipo":str(nombre),
        "Ex_in(kW)":float(Ex_in_kW),
        "Ex_out(kW)":float(Ex_out_kW),
        "ExQ(kW)":float(ExQ_kW),
        "ExW(kW)":float(ExW_kW),
        "ExD(kW)":float(ExD_kW),
    }

def _fila_adiabatica(nombre,Ex_in,Ex_out,ExW):
    ExQ=0.0  # kW
    ExD=float(Ex_in+ExQ+ExW-Ex_out)  # kW
    return _row(nombre,ExD,Ex_in,Ex_out,ExQ,ExW)

def _fila_intercambiador(nombre,Ex_in,Ex_out):
    ExQ=0.0  # kW
    ExW=0.0  # kW
    ExD=float(Ex_in+ExQ+ExW-Ex_out)  # kW
    return _row(nombre,ExD,Ex_in,Ex_out,ExQ,ExW)

def _fila_con_calor(nombre,Ex_in,Ex_out,ExQ):
    ExW=0.0  # kW
    ExD=float(Ex_in+ExQ+ExW-Ex_out)  # kW
    return _row(nombre,ExD,Ex_in,Ex_out,ExQ,ExW)

def _Tb_media_si_existe(parametros,clave,Ta_C,Tb_C,fallback_C):
    if clave in parametros:
        return float(parametros.get(clave))  # °C
    if np.isfinite(Ta_C) and np.isfinite(Tb_C):
        return float(0.5*(float(Ta_C)+float(Tb_C)))  # °C
    return float(fallback_C)  # °C

def calcular_exergia_destruida(parametros,c,df_QW):
    T0_C=float(parametros.get("T0",parametros.get("T0_C",15.0)))  # °C

    eta_gen_efgt=float(parametros.get("eta_generador_efgt",1.0))  # -
    eta_gen_src=float(parametros.get("eta_generador_src",1.0))  # -
    eta_gen_orc=float(parametros.get("eta_generador_orc",1.0))  # -

    Qsolar_in=float(parametros.get("Qsolar_util",parametros.get("Qsolar_util_kW",0.0)))  # kW

    m3=float(parametros.get("m3",0.0))  # kg/s
    mgn=float(parametros.get("mgn",0.0))  # kg/s
    pci_b=float(parametros.get("PCI_bagazo",parametros.get("PCI_bagazo_kJ_kg",0.0)))  # kJ/kg
    pci_g=float(parametros.get("PCI_gn",parametros.get("PCI_gn_kJ_kg",0.0)))  # kJ/kg
    Qfuel_in=float(m3*pci_b+mgn*pci_g)  # kW

    Ex3=float(parametros.get("Ex3",0.0))  # kW
    Exg=float(parametros.get("Exg",0.0))  # kW

    filas=[]

    if 1 in c and 2 in c:
        T1_C=float(_attr_o_clave(c[1],["T"],np.nan))  # °C
        T2_C=float(_attr_o_clave(c[2],["T"],np.nan))  # °C
        Tb_ptc_C=_Tb_media_si_existe(parametros,"T_borde_ptc_C",T1_C,T2_C,max(T1_C,T2_C))  # °C
        Ex_in=_Ex_total(c,2)  # kW
        Ex_out=_Ex_total(c,1)  # kW
        ExQ=_Ex_heat(Qsolar_in,T0_C,Tb_ptc_C)  # kW
        filas.append(_fila_con_calor("PTC_module",Ex_in,Ex_out,ExQ))

    if 1 in c and 2 in c and 12 in c and 13 in c:
        Ex_in=_Ex_total(c,1)+_Ex_total(c,12)  # kW
        Ex_out=_Ex_total(c,2)+_Ex_total(c,13)  # kW
        filas.append(_fila_intercambiador("EFGT_Solar_Preheater_SPH",Ex_in,Ex_out))

    if 4 in c and 5 in c:
        Ex_aire_4_ph=_Ex_ph(c,4)  # kW
        Ex_gases_5_total=_Ex_total(c,5)  # kW
        Ex_in=Ex3+Exg+Ex_aire_4_ph  # kW
        Ex_out=Ex_gases_5_total  # kW
        filas.append(_fila_intercambiador("Combustion_Chamber_BCC",Ex_in,Ex_out))

    if 11 in c and 12 in c:
        Wcomp=_getW(df_QW,"Compresor(11->12)")  # kW
        Ex_in=_Ex_total(c,11)  # kW
        Ex_out=_Ex_total(c,12)  # kW
        ExW=-float(Wcomp)  # kW
        filas.append(_fila_adiabatica("EFGT_Compressor",Ex_in,Ex_out,ExW))

    if 5 in c and 6 in c and 13 in c and 14 in c:
        Ex_in=_Ex_total(c,5)+_Ex_total(c,13)  # kW
        Ex_out=_Ex_total(c,6)+_Ex_total(c,14)  # kW
        filas.append(_fila_intercambiador("EFGT_HTHE",Ex_in,Ex_out))

    if 14 in c and 15 in c:
        Wgt=_getW(df_QW,"Turbina_gas(14->15)")  # kW
        Ex_in=_Ex_total(c,14)  # kW
        Ex_out=_Ex_total(c,15)  # kW
        ExW=-float(Wgt)  # kW
        filas.append(_fila_adiabatica("EFGT_Gas_Turbine_GT",Ex_in,Ex_out,ExW))

    Wcomp=_getW(df_QW,"Compresor(11->12)")  # kW
    Wgt=_getW(df_QW,"Turbina_gas(14->15)")  # kW
    Wm_efgt=float(Wgt+Wcomp)  # kW
    We_efgt=float(eta_gen_efgt*Wm_efgt)  # kW
    Ex_in=float(Wm_efgt)  # kW
    Ex_out=float(We_efgt)  # kW
    ExW=0.0  # kW
    filas.append(_fila_adiabatica("EFGT_Generator",Ex_in,Ex_out,ExW))

    if 15 in c and 16 in c and 21 in c and 24 in c:
        Ex_in=_Ex_total(c,15)+_Ex_total(c,24)  # kW
        Ex_out=_Ex_total(c,16)+_Ex_total(c,21)  # kW
        filas.append(_fila_intercambiador("SRC_Steam_Generator",Ex_in,Ex_out))

    if 21 in c and 22 in c:
        Wts=_getW(df_QW,"Turbina_SRC(21->22)")  # kW
        Ex_in=_Ex_total(c,21)  # kW
        Ex_out=_Ex_total(c,22)  # kW
        ExW=-float(Wts)  # kW
        filas.append(_fila_adiabatica("SRC_Turbine",Ex_in,Ex_out,ExW))

    if 23 in c and 24 in c:
        Wps=_getW(df_QW,"Bomba_SRC(23->24)")  # kW
        Ex_in=_Ex_total(c,23)  # kW
        Ex_out=_Ex_total(c,24)  # kW
        ExW=-float(Wps)  # kW
        filas.append(_fila_adiabatica("SRC_Pump",Ex_in,Ex_out,ExW))

    Wts=_getW(df_QW,"Turbina_SRC(21->22)")  # kW
    Wps=_getW(df_QW,"Bomba_SRC(23->24)")  # kW
    Wm_src=float(Wts+Wps)  # kW
    We_src=float(eta_gen_src*Wm_src)  # kW
    Ex_in=float(Wm_src)  # kW
    Ex_out=float(We_src)  # kW
    ExW=0.0  # kW
    filas.append(_fila_adiabatica("SRC_Generator",Ex_in,Ex_out,ExW))

    if 16 in c and 17 in c and 31 in c and 34 in c:
        Ex_in=_Ex_total(c,16)+_Ex_total(c,34)  # kW
        Ex_out=_Ex_total(c,17)+_Ex_total(c,31)  # kW
        filas.append(_fila_intercambiador("ORC_Vapor_Generator",Ex_in,Ex_out))

    if 31 in c and 32 in c:
        Wto=_getW(df_QW,"Turbina_ORC(31->32)")  # kW
        Ex_in=_Ex_total(c,31)  # kW
        Ex_out=_Ex_total(c,32)  # kW
        ExW=-float(Wto)  # kW
        filas.append(_fila_adiabatica("ORC_Turbine",Ex_in,Ex_out,ExW))

    if 33 in c and 34 in c:
        Wpo=_getW(df_QW,"Bomba_ORC(33->34)")  # kW
        Ex_in=_Ex_total(c,33)  # kW
        Ex_out=_Ex_total(c,34)  # kW
        ExW=-float(Wpo)  # kW
        filas.append(_fila_adiabatica("ORC_Pump",Ex_in,Ex_out,ExW))

    Wto=_getW(df_QW,"Turbina_ORC(31->32)")  # kW
    Wpo=_getW(df_QW,"Bomba_ORC(33->34)")  # kW
    Wm_orc=float(Wto+Wpo)  # kW
    We_orc=float(eta_gen_orc*Wm_orc)  # kW
    Ex_in=float(Wm_orc)  # kW
    Ex_out=float(We_orc)  # kW
    ExW=0.0  # kW
    filas.append(_fila_adiabatica("ORC_Generator",Ex_in,Ex_out,ExW))

    if all(n in c for n in [67,68,71,72]):
        Ex_in=_Ex_total(c,67)+_Ex_total(c,71)  # kW
        Ex_out=_Ex_total(c,68)+_Ex_total(c,72)  # kW
        filas.append(_fila_intercambiador("SEAC_Condenser",Ex_in,Ex_out))

    if all(n in c for n in [66,70,61,73,74]):
        Ex_in=_Ex_total(c,66)+_Ex_total(c,70)+_Ex_total(c,73)  # kW
        Ex_out=_Ex_total(c,61)+_Ex_total(c,74)  # kW
        filas.append(_fila_intercambiador("SEAC_Absorber",Ex_in,Ex_out))

    if all(n in c for n in [61,62]):
        Wbp=_getW(df_QW,"Bomba_SEAC(61->62)")  # kW
        Ex_in=_Ex_total(c,61)  # kW
        Ex_out=_Ex_total(c,62)  # kW
        ExW=-float(Wbp)  # kW
        filas.append(_fila_adiabatica("SEAC_Pump",Ex_in,Ex_out,ExW))

    if all(n in c for n in [64,63,67]):
        Qg=float(parametros.get("Qg",parametros.get("Q_generador_seac",0.0)))  # kW
        T63_C=float(parametros.get("T63",float(c[63].T)))  # °C
        T64_C=float(parametros.get("T64",float(c[64].T)))  # °C
        Tb_gen_C=_Tb_media_si_existe(parametros,"Tb_gen_seac",T63_C,T64_C,80.0)  # °C
        Ex_in=_Ex_total(c,64)  # kW
        Ex_out=_Ex_total(c,63)+_Ex_total(c,67)  # kW
        ExQ=_Ex_heat(Qg,T0_C,Tb_gen_C)  # kW
        filas.append(_fila_con_calor("SEAC_Generator",Ex_in,Ex_out,ExQ))

    if all(n in c for n in [69,70]):
        Qev=float(parametros.get("Q_evaporador_seac",parametros.get("Q_ev_seac",0.0)))  # kW
        Tb_evap_C=float(parametros.get("Tb_evap_seac",parametros.get("T69",float(c[69].T))))  # °C
        Ex_in=_Ex_total(c,69)  # kW
        Ex_out=_Ex_total(c,70)  # kW
        ExQ=_Ex_heat(Qev,T0_C,Tb_evap_C)  # kW
        filas.append(_fila_con_calor("SEAC_Evaporator",Ex_in,Ex_out,ExQ))

    if all(n in c for n in [6,7,43,45,58,59]):
        Ex_in=_Ex_total(c,6)+_Ex_total(c,43)  # kW
        Ex_out=_Ex_total(c,7)+_Ex_total(c,45)+_Ex_total(c,58)+_Ex_total(c,59)  # kW
        filas.append(_fila_intercambiador("MED_con_calentador_BCC",Ex_in,Ex_out))

    df=pd.DataFrame(filas)

    tot_ExD=float(np.nansum(df["ExD(kW)"].values))  # kW
    tot_ExD=tot_ExD if np.isfinite(tot_ExD) else float("nan")

    resumen=pd.DataFrame([
        {"Indicador":"T0(°C)","Valor":float(T0_C)},
        {"Indicador":"Qsolar_in(kW)","Valor":float(Qsolar_in)},
        {"Indicador":"Qfuel_in(kW)","Valor":float(Qfuel_in)},
        {"Indicador":"Ex_fuel_in(kW)","Valor":float(Ex3+Exg)},
        {"Indicador":"ExD_total_sum(kW)","Valor":float(tot_ExD)},
    ])

    print(df.to_string(index=False))
    print(resumen.to_string(index=False))
    return df,resumen

df_exd,resumen_exd=calcular_exergia_destruida(parametros_base,c_total,df_QW)
```

```python
import numpy as np
import pandas as pd

def _safe_div(num,den):
    num=float(num)
    den=float(den)
    return float("nan") if abs(den)<=1e-12 else num/den

def _Ex_hot(Q_kW,T0_C,Tb_C):
    T0_K=float(T0_C)+273.15  # K
    Tb_K=float(Tb_C)+273.15  # K
    Q_kW=float(Q_kW)
    if abs(Q_kW)<=1e-12:
        return 0.0  # kW
    if Tb_K<=1e-9:
        return 0.0  # kW
    return Q_kW*(1.0-T0_K/Tb_K)  # kW

def _Ex_cooling(Qc_kW,T0_C,Tc_C):
    T0_K=float(T0_C)+273.15  # K
    Tc_K=float(Tc_C)+273.15  # K
    Qc_kW=float(Qc_kW)
    if abs(Qc_kW)<=1e-12:
        return 0.0  # kW
    if Tc_K<=1e-9:
        return 0.0  # kW
    return Qc_kW*(T0_K/Tc_K-1.0)  # kW

def _attr_o_clave(x,nombres,default=None):
    for nombre in nombres:
        if hasattr(x,nombre):
            try:
                return float(getattr(x,nombre))
            except:
                pass
        if isinstance(x,dict) and nombre in x:
            try:
                return float(x[nombre])
            except:
                pass
    return default

def _Ex_total(c,n):
    if c is None or n not in c:
        return 0.0  # kW
    x=c[n]
    v_total=_attr_o_clave(x,["Ex"],None)
    v_fisica=_attr_o_clave(x,["Ex_fisica","ex_fisica","Ex_ph","ex_ph"],None)
    v_quimica=_attr_o_clave(x,["Ex_quimica","ex_quimica","Ex_ch","ex_ch"],None)
    if (v_fisica is not None) or (v_quimica is not None):
        return float((0.0 if v_fisica is None else v_fisica)+(0.0 if v_quimica is None else v_quimica))  # kW
    if v_total is not None:
        return float(v_total)  # kW
    return 0.0  # kW

def calcular_eficiencias_exergeticas(parametros,df_exd,df_QW,c_total=None):
    T0_C=float(parametros.get("T0",parametros.get("T0_C",15.0)))  # °C

    eta_gen_efgt=float(parametros.get("eta_generador_efgt",1.0))  # -
    eta_gen_src=float(parametros.get("eta_generador_src",1.0))  # -
    eta_gen_orc=float(parametros.get("eta_generador_orc",1.0))  # -

    Qsolar_in=float(parametros.get("Qsolar_util",parametros.get("Qsolar_util_kW",0.0)))  # kW

    m3=float(parametros.get("m3",0.0))  # kg/s
    mgn=float(parametros.get("mgn",0.0))  # kg/s
    pci_b=float(parametros.get("PCI_bagazo",parametros.get("PCI_bagazo_kJ_kg",0.0)))  # kJ/kg
    pci_g=float(parametros.get("PCI_gn",parametros.get("PCI_gn_kJ_kg",0.0)))  # kJ/kg
    Qfuel_in=float(m3*pci_b+mgn*pci_g)  # kW

    Ex_fuel_in=float(parametros.get("Ex3",0.0))+float(parametros.get("Exg",0.0))  # kW

    Tb_ptc_C=float(parametros.get("T_borde_ptc_C",parametros.get("Tb_solar_C",parametros.get("Tb_solar",390.0))))  # °C
    Ex_aporte_solar_util=_Ex_hot(Qsolar_in,T0_C,Tb_ptc_C)  # kW

    Ex_in_total=float(Ex_fuel_in+Ex_aporte_solar_util)  # kW

    def _getW(nombre):
        v=df_QW.loc[df_QW["Subunidad"]==nombre,"W(kW)"]
        return float(v.iloc[0]) if len(v) else 0.0  # kW

    def _getQ(nombre):
        v=df_QW.loc[df_QW["Subunidad"]==nombre,"Q(kW)"]
        return float(v.iloc[0]) if len(v) else 0.0  # kW

    W_comp=abs(_getW("Compresor(11->12)"))  # kW
    W_tg=abs(_getW("Turbina_gas(14->15)"))  # kW
    Wm_efgt=float(W_tg-W_comp)  # kW
    W_efgt_e=float(eta_gen_efgt*Wm_efgt)  # kW

    W_ts=abs(_getW("Turbina_SRC(21->22)"))  # kW
    W_ps=abs(_getW("Bomba_SRC(23->24)"))  # kW
    Wm_src=float(W_ts-W_ps)  # kW
    W_src_e=float(eta_gen_src*Wm_src)  # kW

    W_to=abs(_getW("Turbina_ORC(31->32)"))  # kW
    W_po=abs(_getW("Bomba_ORC(33->34)"))  # kW
    Wm_orc=float(W_to-W_po)  # kW
    W_orc_e=float(eta_gen_orc*Wm_orc)  # kW

    W_aux_seac=abs(_getW("Bomba_SEAC(61->62)"))  # kW

    W_neta=float(W_efgt_e+W_src_e+W_orc_e-W_aux_seac)  # kW

    Q_dhu=_getQ("DHU(28->29)")  # kW
    if abs(Q_dhu)<=1e-12:
        Q_dhu=float(parametros.get("Qdhu",parametros.get("Q_calefaccion",parametros.get("Q_calefaccion_obj",0.0))))  # kW

    T28_C=float(parametros.get("T28",np.nan))  # °C
    T29_C=float(parametros.get("T29",np.nan))  # °C
    if np.isfinite(T28_C) and np.isfinite(T29_C):
        Tb_dhu_C=float(parametros.get("Tb_dhu_C",0.5*(T28_C+T29_C)))  # °C
    else:
        Tb_dhu_C=float(parametros.get("Tb_dhu_C",parametros.get("T29",40.0)))  # °C
    Ex_heat_dhu=_Ex_hot(Q_dhu,T0_C,Tb_dhu_C)  # kW

    Q_ev_seac=float(parametros.get("Q_evaporador_seac",parametros.get("Q_ev_seac",0.0)))  # kW
    if abs(Q_ev_seac)<=1e-12:
        Q_ev_seac=_getQ("SEAC_Evaporator")  # kW
    if abs(Q_ev_seac)<=1e-12:
        Q_ev_seac=_getQ("Evap_SEAC(69->70)")  # kW

    T75_C=float(parametros.get("T75",np.nan))  # °C
    T76_C=float(parametros.get("T76",np.nan))  # °C
    if np.isfinite(T75_C) and np.isfinite(T76_C):
        Tc_ev_seac_C=float(parametros.get("Tc_ev_seac_C",0.5*(T75_C+T76_C)))  # °C
    else:
        Tc_ev_seac_C=float(parametros.get("Tc_ev_seac_C",parametros.get("T69",10.0)))  # °C
    Ex_cool_seac=_Ex_cooling(Q_ev_seac,T0_C,Tc_ev_seac_C)  # kW

    Ex_agua_potable=_Ex_total(c_total,59)  # kW

    ExD_total=float(np.nansum(df_exd["ExD(kW)"].values))  # kW

    eta_ex_el=_safe_div(W_neta,Ex_in_total)  # -
    eta_ex_cal=_safe_div(Ex_heat_dhu,Ex_in_total)  # -
    eta_ex_frio=_safe_div(Ex_cool_seac,Ex_in_total)  # -
    eta_ex_agua=_safe_div(Ex_agua_potable,Ex_in_total)  # -
    eta_ex_total_util=_safe_div(W_neta+Ex_heat_dhu+Ex_cool_seac+Ex_agua_potable,Ex_in_total)  # -

    cierre_parcial=_safe_div(ExD_total+(W_neta+Ex_heat_dhu+Ex_cool_seac+Ex_agua_potable),Ex_in_total)  # -

    resumen=pd.DataFrame([
        {"Indicador":"T0(°C)","Valor":T0_C},
        {"Indicador":"Qsolar_in(kW)","Valor":Qsolar_in},
        {"Indicador":"Tb_ptc(°C)","Valor":Tb_ptc_C},
        {"Indicador":"Ex_aporte_solar_util(kW)","Valor":Ex_aporte_solar_util},
        {"Indicador":"Qfuel_in(kW)","Valor":Qfuel_in},
        {"Indicador":"Ex_fuel_in(kW)","Valor":Ex_fuel_in},
        {"Indicador":"Ex_in_total(kW)","Valor":Ex_in_total},
        {"Indicador":"W_neta(kW)","Valor":W_neta},
        {"Indicador":"Q_DHU(kW)","Valor":Q_dhu},
        {"Indicador":"Tb_DHU(°C)","Valor":Tb_dhu_C},
        {"Indicador":"Ex_calefaccion(kW)","Valor":Ex_heat_dhu},
        {"Indicador":"Q_ev_SEAC(kW)","Valor":Q_ev_seac},
        {"Indicador":"Tc_ev_SEAC(°C)","Valor":Tc_ev_seac_C},
        {"Indicador":"Ex_enfriamiento(kW)","Valor":Ex_cool_seac},
        {"Indicador":"Ex_agua_potable(kW)","Valor":Ex_agua_potable},
        {"Indicador":"ExD_total_sum(kW)","Valor":ExD_total},
        {"Indicador":"eta_ex_el(-)","Valor":eta_ex_el},
        {"Indicador":"eta_ex_calefaccion(-)","Valor":eta_ex_cal},
        {"Indicador":"eta_ex_enfriamiento(-)","Valor":eta_ex_frio},
        {"Indicador":"eta_ex_agua(-)","Valor":eta_ex_agua},
        {"Indicador":"eta_ex_total_util(-)","Valor":eta_ex_total_util},
        {"Indicador":"cierre_parcial_exergia(-)","Valor":cierre_parcial},
    ])

    print(resumen.to_string(index=False))
    return resumen

df_etas_ex=calcular_eficiencias_exergeticas(parametros_base,df_exd,df_QW,c_total=c_total)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def _inferir_subunidad_por_nodo(n):
    n=int(n)
    if 1<=n<=2:
        return "PTC"
    if 3<=n<=7:
        return "BCC"
    if 11<=n<=17:
        return "EFGT"
    if 21<=n<=27:
        return "SRC"
    if 28<=n<=29:
        return "DHU"
    if 31<=n<=40:
        return "ORC"
    if 41<=n<=60:
        return "MED"
    if 61<=n<=76:
        return "SEAC"
    return "OTROS"

def _leer_atributo_corriente(co,nombre,default=np.nan):
    if hasattr(co,nombre):
        try:
            return getattr(co,nombre)
        except:
            return default
    if isinstance(co,dict) and nombre in co:
        try:
            return co[nombre]
        except:
            return default
    return default

def _construir_df_total_desde_corrientes(c_total):
    filas=[]
    for n,co in c_total.items():
        n_int=int(n)

        bloque=_leer_atributo_corriente(co,"bloque",None)
        if bloque is None:
            bloque=_leer_atributo_corriente(co,"b",None)
        if bloque is None:
            bloque=""

        ex_total=_leer_atributo_corriente(co,"Ex",np.nan)

        ex_fis=_leer_atributo_corriente(co,"Ex_fis",None)
        if ex_fis is None or pd.isna(ex_fis):
            ex_fis=_leer_atributo_corriente(co,"ex_fis",None)
        if ex_fis is None or pd.isna(ex_fis):
            ex_fis=_leer_atributo_corriente(co,"Ex_fisica",None)
        if ex_fis is None or pd.isna(ex_fis):
            ex_fis=_leer_atributo_corriente(co,"ex_fisica",None)
        if ex_fis is None or pd.isna(ex_fis):
            ex_fis=_leer_atributo_corriente(co,"Ex_ph",None)
        if ex_fis is None or pd.isna(ex_fis):
            ex_fis=_leer_atributo_corriente(co,"ex_ph",None)

        ex_quim=_leer_atributo_corriente(co,"Ex_quim",None)
        if ex_quim is None or pd.isna(ex_quim):
            ex_quim=_leer_atributo_corriente(co,"ex_quim",None)
        if ex_quim is None or pd.isna(ex_quim):
            ex_quim=_leer_atributo_corriente(co,"Ex_quimica",None)
        if ex_quim is None or pd.isna(ex_quim):
            ex_quim=_leer_atributo_corriente(co,"ex_quimica",None)
        if ex_quim is None or pd.isna(ex_quim):
            ex_quim=_leer_atributo_corriente(co,"Ex_ch",None)
        if ex_quim is None or pd.isna(ex_quim):
            ex_quim=_leer_atributo_corriente(co,"ex_ch",None)

        ex_fis_val=float(ex_fis) if ex_fis is not None and pd.notna(ex_fis) else np.nan
        ex_quim_val=float(ex_quim) if ex_quim is not None and pd.notna(ex_quim) else np.nan
        ex_total_val=float(ex_total) if pd.notna(ex_total) else np.nan

        if pd.isna(ex_total_val):
            ex_total_val=(0.0 if pd.isna(ex_fis_val) else ex_fis_val)+(0.0 if pd.isna(ex_quim_val) else ex_quim_val)

        filas.append({
            "corriente":n_int,
            "bloque":str(bloque),
            "ex_kW":ex_total_val,
            "ex_fis_kW":ex_fis_val,
            "ex_quim_kW":ex_quim_val,
        })

    df=pd.DataFrame(filas)
    df["corriente"]=pd.to_numeric(df["corriente"],errors="coerce")
    df["bloque"]=df["bloque"].astype(str).fillna("")
    return df

def _asegurar_df_total(df_total=None,c_total=None):
    if isinstance(df_total,pd.DataFrame):
        return df_total.copy(deep=True)

    if "df_total" in globals() and isinstance(globals()["df_total"],pd.DataFrame):
        return globals()["df_total"].copy(deep=True)

    if c_total is None and "c_total" in globals() and isinstance(globals()["c_total"],dict):
        c_total=globals()["c_total"]

    if isinstance(c_total,dict) and len(c_total)>0:
        return _construir_df_total_desde_corrientes(c_total)

    raise NameError("df_total no existe como DataFrame y no se pudo construir desde c_total")

def _requerir_columnas(df,cols,nombre="DataFrame"):
    faltantes=[c for c in cols if c not in df.columns]
    if faltantes:
        raise KeyError(f"Faltan columnas en {nombre}: {faltantes}")

def _agregar_ex_total(df,col_ex_fis="ex_fis_kW",col_ex_quim="ex_quim_kW",col_ex="ex_kW"):
    df=df.copy()

    if col_ex not in df.columns:
        df[col_ex]=np.nan
    if col_ex_fis not in df.columns:
        df[col_ex_fis]=np.nan
    if col_ex_quim not in df.columns:
        df[col_ex_quim]=np.nan

    df[col_ex]=pd.to_numeric(df[col_ex],errors="coerce")
    df[col_ex_fis]=pd.to_numeric(df[col_ex_fis],errors="coerce")
    df[col_ex_quim]=pd.to_numeric(df[col_ex_quim],errors="coerce")

    ex_reconstruida=df[col_ex_fis].fillna(0.0)+df[col_ex_quim].fillna(0.0)
    df[col_ex]=df[col_ex].where(df[col_ex].notna(),ex_reconstruida)

    return df

def _vector_ex_para_plot(x,modo):
    x=pd.to_numeric(x,errors="coerce").fillna(0.0)
    if modo=="abs":
        return x.abs()
    if modo=="clip0":
        return x.clip(lower=0.0)
    if modo=="raw":
        return x
    raise ValueError("modo inválido. Usa 'abs', 'clip0' o 'raw'")

def _agrupar_otros(serie,min_pct_otros):
    if serie.empty:
        return serie
    total=float(serie.sum())
    if total<=0:
        return serie
    pct=100.0*serie/total
    chicos=pct[pct<float(min_pct_otros)].index
    if len(chicos)==0:
        return serie
    otros_val=float(serie.loc[chicos].sum())
    serie2=serie.drop(index=chicos).copy()
    serie2.loc["Otros"]=otros_val
    return serie2.sort_values(ascending=False)

def _pastel_con_leyenda(serie_kw,titulo,cmap_name="viridis",min_pct_otros=0.0,unidades="kW",n_decimales_kw=2):
    serie_kw=pd.to_numeric(serie_kw,errors="coerce").fillna(0.0)
    serie_kw=serie_kw[serie_kw>0].sort_values(ascending=False)

    if min_pct_otros and min_pct_otros>0:
        serie_kw=_agrupar_otros(serie_kw,min_pct_otros=min_pct_otros)

    if len(serie_kw)<=1:
        return

    fig,ax=plt.subplots(figsize=(8,5))
    cmap=plt.get_cmap(cmap_name)
    colores=cmap(np.linspace(0.1,0.9,len(serie_kw)))

    wedges,_=ax.pie(
        serie_kw.values,
        startangle=90,
        colors=colores,
        wedgeprops={"edgecolor":"white"},
    )
    ax.axis("equal")
    ax.set_title(titulo,fontsize=12,fontweight="bold",pad=20)

    total=float(serie_kw.sum())
    etiquetas=[]
    for nombre,valor in serie_kw.items():
        etiquetas.append(f"{nombre} — {valor:.{n_decimales_kw}f} {unidades} ({valor/total*100:.1f}%)")

    ax.legend(
        wedges,
        etiquetas,
        title="Componentes",
        loc="center left",
        bbox_to_anchor=(1.0,0.5),
        fontsize=9,
        frameon=True,
    )

    plt.subplots_adjust(right=0.7)
    plt.tight_layout()
    plt.show()

def _barras_horizontales(serie_kw,titulo,xlabel="Exergía (kW)",n_max=12,min_pct_otros=0.0,cmap_name="viridis",unidades="kW",n_decimales_kw=2):
    serie_kw=pd.to_numeric(serie_kw,errors="coerce").fillna(0.0)
    serie_kw=serie_kw[serie_kw>0].sort_values(ascending=False)

    if min_pct_otros and min_pct_otros>0:
        serie_kw=_agrupar_otros(serie_kw,min_pct_otros=min_pct_otros)

    if n_max and len(serie_kw)>int(n_max):
        top=serie_kw.iloc[:int(n_max)]
        resto=serie_kw.iloc[int(n_max):].sum()
        serie_kw=top.copy()
        if resto>0:
            serie_kw.loc["Otros"]=float(resto)

    serie_kw=serie_kw.sort_values(ascending=True)
    if len(serie_kw)<=1:
        return

    fig,ax=plt.subplots(figsize=(8,5))
    cmap=plt.get_cmap(cmap_name)
    colores=cmap(np.linspace(0.1,0.9,len(serie_kw)))

    ax.barh(serie_kw.index.astype(str),serie_kw.values,color=colores)
    ax.set_title(titulo,fontsize=12,fontweight="bold",pad=10)
    ax.set_xlabel(xlabel)

    xmax=float(serie_kw.max())
    dx=max(0.01*xmax,1e-9)
    for y,(nombre,valor) in enumerate(serie_kw.items()):
        ax.text(valor+dx,y,f"{valor:.{n_decimales_kw}f} {unidades}",va="center",fontsize=9)

    plt.tight_layout()
    plt.show()

def _clasificar_subunidad(df,mapa_subunidad):
    df=df.copy()
    _requerir_columnas(df,["bloque","corriente"],nombre="df_total")

    bloque=df["bloque"].astype(str).str.strip()
    sub=bloque.map(mapa_subunidad)

    sub_directa=bloque.where(bloque.isin(set(mapa_subunidad.values())),other=np.nan)
    sub=sub.fillna(sub_directa)

    sub=sub.fillna(pd.to_numeric(df["corriente"],errors="coerce").map(_inferir_subunidad_por_nodo))
    sub=sub.fillna("OTROS")

    df["subunidad"]=sub.astype(str)
    return df

def _global_por_subunidad(df,modo,col_ex):
    df=df.copy()
    _requerir_columnas(df,["subunidad",col_ex],nombre="df_total")
    df["_ex_plot_"]=_vector_ex_para_plot(df[col_ex],modo)
    df_g=df.groupby("subunidad",as_index=False)["_ex_plot_"].sum().rename(columns={"_ex_plot_":"Ex_kW"})
    df_g=df_g[df_g["Ex_kW"]>0].copy()
    total=df_g["Ex_kW"].sum()
    df_g["porcentaje"]=100.0*df_g["Ex_kW"]/total if total>0 else 0.0
    df_g=df_g.sort_values("Ex_kW",ascending=False).reset_index(drop=True)
    return df_g

def _por_equipo_en_subunidad(df,subunidad,mapa_equipo,modo,col_ex):
    df_s=df[df["subunidad"]==subunidad].copy()
    if df_s.empty:
        return pd.DataFrame(columns=["subunidad","equipo","Ex_kW","porcentaje"])

    _requerir_columnas(df_s,["corriente",col_ex],nombre=f"df_total[{subunidad}]")

    corr_num=pd.to_numeric(df_s["corriente"],errors="coerce")
    df_s["equipo"]=corr_num.map(mapa_equipo)

    if df_s["equipo"].isna().all():
        df_s["equipo"]=df_s["corriente"].astype(str).map(mapa_equipo)

    df_s=df_s.dropna(subset=["equipo"]).copy()
    if df_s.empty:
        return pd.DataFrame(columns=["subunidad","equipo","Ex_kW","porcentaje"])

    df_s["_ex_plot_"]=_vector_ex_para_plot(df_s[col_ex],modo)
    df_e=df_s.groupby("equipo",as_index=False)["_ex_plot_"].sum().rename(columns={"_ex_plot_":"Ex_kW"})
    df_e=df_e[df_e["Ex_kW"]>0].copy()
    total=df_e["Ex_kW"].sum()
    df_e["porcentaje"]=100.0*df_e["Ex_kW"]/total if total>0 else 0.0
    df_e.insert(0,"subunidad",subunidad)
    df_e=df_e.sort_values("Ex_kW",ascending=False).reset_index(drop=True)
    return df_e

def correr_graficas_exergia(
    df_total=None,
    c_total=None,
    modo="abs",
    col_ex="ex_kW",
    col_ex_fis="ex_fis_kW",
    col_ex_quim="ex_quim_kW",
    mapa_subunidad=None,
    mapas_equipo_por_subunidad=None,
    min_pct_otros=2.0,
    mostrar_pastel=True,
    mostrar_barras=True,
    cmap_global="viridis",
    paletas=None,
    n_max_barras=12,
    incluir_subunidades_cero=True,
):
    df=_asegurar_df_total(df_total,c_total=c_total)
    _requerir_columnas(df,["corriente"],nombre="df_total")

    if "bloque" not in df.columns:
        df["bloque"]=""

    df=_agregar_ex_total(df,col_ex_fis=col_ex_fis,col_ex_quim=col_ex_quim,col_ex=col_ex)

    if mapa_subunidad is None:
        mapa_subunidad={
            "corrientes_cc":"BCC",
            "corrientes_ptc":"PTC",
            "corrientes_efgt":"EFGT",
            "corrientes_src":"SRC",
            "corrientes_orc":"ORC",
            "corrientes_med":"MED",
            "corrientes_61_76":"SEAC",
            "corrientes_dhu_28_29":"DHU",
            "PTC":"PTC",
            "BCC":"BCC",
            "EFGT":"EFGT",
            "SRC":"SRC",
            "ORC":"ORC",
            "MED":"MED",
            "SEAC":"SEAC",
            "DHU":"DHU",
        }

    if mapas_equipo_por_subunidad is None:
        mapas_equipo_por_subunidad={
            "PTC":{
                1:"Therminol caliente hacia SPH",
                2:"Therminol frío hacia colector",
            },
            "BCC":{
                3:"Bagazo/combustible",
                4:"Aire",
                5:"Gases combustión",
                6:"Gases tras HTHE",
                7:"Chimenea",
            },
            "EFGT":{
                11:"Aire ambiente",
                12:"Compresor",
                13:"Precalentador solar",
                14:"HTHE",
                15:"Turbina gas",
                16:"Salida a SRC/ORC",
                17:"Escape final",
            },
            "SRC":{
                21:"Entrada turbina",
                22:"Salida turbina",
                23:"Condensador",
                24:"Bomba",
            },
            "DHU":{
                28:"Entrada agua calefacción",
                29:"Salida agua calefacción",
            },
            "ORC":{
                31:"Entrada turbina",
                32:"Salida turbina",
                33:"Condensador",
                34:"Bomba",
                38:"Agua enfriamiento in",
                39:"Agua enfriamiento out",
            },
            "MED":{
                41:"Agua caliente in",
                42:"Agua caliente out",
                43:"Agua de mar in",
                44:"Agua de mar precalentada",
                45:"Derivación agua de mar",
                46:"Entrada a efectos",
                47:"Vapor efecto 1",
                48:"Salmuera efecto 1",
                49:"Destilado efecto 1",
                50:"Vapor efecto 2",
                51:"Salmuera efecto 2",
                52:"Destilado efecto 2",
                53:"Vapor efecto 3",
                54:"Salmuera efecto 3",
                55:"Destilado efecto 3",
                56:"Vapor efecto 4",
                57:"Destilado efecto 4",
                58:"Salmuera final",
                59:"Destilado total",
            },
            "SEAC":{
                61:"Absorbedor salida débil",
                62:"Bomba solución",
                63:"Salida HX a generador",
                64:"Generador salida fuerte",
                65:"HX solución fuerte",
                66:"Válvula solución",
                67:"Generador/condensador vapor",
                68:"Condensado",
                69:"Válvula refrigerante",
                70:"Evaporador vapor",
                71:"Agua enfriamiento condensador in",
                72:"Agua enfriamiento condensador out",
                73:"Agua enfriamiento absorbedor in",
                74:"Agua enfriamiento absorbedor out",
                75:"Agua a enfriar in",
                76:"Agua enfriada out",
            },
        }

    if paletas is None:
        paletas=["viridis","plasma","magma","cividis","cool","cubehelix"]

    df=_clasificar_subunidad(df,mapa_subunidad)
    df_global=_global_por_subunidad(df,modo=modo,col_ex=col_ex)

    if incluir_subunidades_cero:
        todas=["PTC","BCC","EFGT","SRC","ORC","MED","SEAC","DHU"]
        presentes=set(df_global["subunidad"].astype(str).tolist())
        faltan=[s for s in todas if s not in presentes]
        if faltan:
            df_zero=pd.DataFrame({"subunidad":faltan,"Ex_kW":0.0,"porcentaje":0.0})
            df_global=pd.concat([df_global,df_zero],ignore_index=True)
            df_global=df_global.sort_values("Ex_kW",ascending=False).reset_index(drop=True)

    if not df_global.empty:
        serie_global=df_global.set_index("subunidad")["Ex_kW"]
        if mostrar_pastel:
            _pastel_con_leyenda(
                serie_global,
                "Distribución de exergía por subunidad",
                cmap_name=cmap_global,
                min_pct_otros=min_pct_otros,
            )
        if mostrar_barras:
            _barras_horizontales(
                serie_global,
                "Distribución de exergía por subunidad (barras)",
                cmap_name=cmap_global,
                min_pct_otros=min_pct_otros,
                n_max=n_max_barras,
            )

    lista_df_sub=[]
    for i,(sub,mapa_eq) in enumerate(mapas_equipo_por_subunidad.items()):
        df_sub=_por_equipo_en_subunidad(df,sub,mapa_eq,modo=modo,col_ex=col_ex)
        if df_sub.empty:
            continue
        lista_df_sub.append(df_sub)
        serie_eq=df_sub.set_index("equipo")["Ex_kW"]
        cmap=paletas[i%len(paletas)]
        if mostrar_pastel:
            _pastel_con_leyenda(
                serie_eq,
                f"{sub}: distribución por equipo",
                cmap_name=cmap,
                min_pct_otros=min_pct_otros,
            )
        if mostrar_barras:
            _barras_horizontales(
                serie_eq,
                f"{sub}: distribución por equipo (barras)",
                cmap_name=cmap,
                min_pct_otros=min_pct_otros,
                n_max=n_max_barras,
            )

    if len(lista_df_sub)>0:
        df_por_equipo=pd.concat(lista_df_sub,ignore_index=True)
    else:
        df_por_equipo=pd.DataFrame(columns=["subunidad","equipo","Ex_kW","porcentaje"])

    return df_global,df_por_equipo,df

c_total={**corr_ptc,**corr_bcc,**corr_efgt,**corr_src,**corr_orc,**corr_med,**corr_seac,**corr_dhu}

df_global_sub,df_por_equipo,df_total_auto=correr_graficas_exergia(
    df_total=None,
    c_total=c_total,
    modo="abs",
    col_ex="ex_kW",
    mostrar_pastel=True,
    mostrar_barras=True,
    min_pct_otros=2.0,
    n_max_barras=12,
)
```

```python
import pandas as pd
import numpy as np

def _speco_tomar(df,candidatas,nueva):
    for c in candidatas:
        if c in df.columns:
            df=df.rename(columns={c:nueva})
            return df
    return df

def _speco_subunidad(n):
    n=int(n)
    if 1<=n<=2:
        return "PTC"
    if 3<=n<=7:
        return "BCC"
    if 11<=n<=17:
        return "EFGT"
    if 21<=n<=24:
        return "SRC"
    if 28<=n<=29:
        return "DHU"
    if 31<=n<=39:
        return "ORC"
    if 41<=n<=59:
        return "MED"
    if 61<=n<=76:
        return "SEAC"
    return "OTROS"

df_speco_corrientes=_construir_df_total_desde_corrientes(c_total).copy()

df_speco_corrientes=_speco_tomar(df_speco_corrientes,["corriente"],"n")
df_speco_corrientes=_speco_tomar(df_speco_corrientes,["fluido","Fluido","f"],"fluido")
df_speco_corrientes=_speco_tomar(df_speco_corrientes,["bloque","Bloque","b"],"bloque")
df_speco_corrientes=_speco_tomar(df_speco_corrientes,["subunidad","Subunidad"],"subunidad")
df_speco_corrientes=_speco_tomar(df_speco_corrientes,["P","P(kPa)","P_kPa"],"P")
df_speco_corrientes=_speco_tomar(df_speco_corrientes,["T","T(°C)","T_C"],"T")
df_speco_corrientes=_speco_tomar(df_speco_corrientes,["m","m(kg/s)","m_kg_s"],"m")
df_speco_corrientes=_speco_tomar(df_speco_corrientes,["h","h(kJ/kg)","h_kJ_kg"],"h")
df_speco_corrientes=_speco_tomar(df_speco_corrientes,["s","s(kJ/kg K)","s_kJ_kgK"],"s")
df_speco_corrientes=_speco_tomar(df_speco_corrientes,["Ex","ex_kW","Ex(kW)"],"Ex")
df_speco_corrientes=_speco_tomar(df_speco_corrientes,["Ex_fisica","ex_fis_kW","Ex_fisica(kW)"],"Ex_fisica")
df_speco_corrientes=_speco_tomar(df_speco_corrientes,["Ex_quimica","ex_quim_kW","Ex_quimica(kW)"],"Ex_quimica")

if "n" not in df_speco_corrientes.columns:
    raise ValueError(f"No aparece la columna de nodo/corriente. Columnas: {list(df_speco_corrientes.columns)}")

if "Ex" not in df_speco_corrientes.columns:
    raise ValueError(f"Falta la columna Ex. Columnas: {list(df_speco_corrientes.columns)}")

if "fluido" not in df_speco_corrientes.columns:
    df_speco_corrientes["fluido"]=""

if "bloque" not in df_speco_corrientes.columns:
    df_speco_corrientes["bloque"]=""

if "subunidad" not in df_speco_corrientes.columns:
    df_speco_corrientes["subunidad"]=np.nan

for c in ["n","P","T","m","h","s","Ex","Ex_fisica","Ex_quimica"]:
    if c not in df_speco_corrientes.columns:
        df_speco_corrientes[c]=np.nan
    df_speco_corrientes[c]=pd.to_numeric(df_speco_corrientes[c],errors="coerce")

df_speco_corrientes=df_speco_corrientes.dropna(subset=["n"]).copy()
df_speco_corrientes["n"]=df_speco_corrientes["n"].astype(int)

df_speco_corrientes["fluido"]=df_speco_corrientes["fluido"].fillna("").astype(str)
df_speco_corrientes["bloque"]=df_speco_corrientes["bloque"].fillna("").astype(str)
df_speco_corrientes["subunidad"]=df_speco_corrientes["subunidad"].fillna("").astype(str)

mask_sub=df_speco_corrientes["subunidad"].str.strip()==""
df_speco_corrientes.loc[mask_sub,"subunidad"]=df_speco_corrientes.loc[mask_sub,"n"].apply(_speco_subunidad)

mask_bloque=df_speco_corrientes["bloque"].str.strip()==""
df_speco_corrientes.loc[mask_bloque,"bloque"]=df_speco_corrientes.loc[mask_bloque,"subunidad"]

df_speco_corrientes["Ex_quimica"]=df_speco_corrientes["Ex_quimica"].fillna(0.0)

mask_fis=df_speco_corrientes["Ex_fisica"].isna()
df_speco_corrientes.loc[mask_fis,"Ex_fisica"]=df_speco_corrientes.loc[mask_fis,"Ex"]-df_speco_corrientes.loc[mask_fis,"Ex_quimica"]

df_speco_corrientes["error_Ex"]=df_speco_corrientes["Ex"]-(df_speco_corrientes["Ex_fisica"]+df_speco_corrientes["Ex_quimica"])
if (df_speco_corrientes["error_Ex"].abs()>1e-6).any():
    raise ValueError("Ex != Ex_fisica + Ex_quimica en una o mas corrientes")

df_speco_corrientes["usa_ex_quimica"]=df_speco_corrientes["Ex_quimica"].abs()>1e-9
df_speco_corrientes["variable_costo"]=df_speco_corrientes["n"].apply(lambda n:f"c{int(n)}")

df_speco_corrientes=df_speco_corrientes[[
    "n","fluido","bloque","subunidad","P","T","m","h","s",
    "Ex","Ex_fisica","Ex_quimica","usa_ex_quimica","variable_costo"
]].sort_values("n").reset_index(drop=True)

ex_corriente=df_speco_corrientes.set_index("n")["Ex"].to_dict()
ex_fisica_corriente=df_speco_corrientes.set_index("n")["Ex_fisica"].to_dict()
ex_quimica_corriente=df_speco_corrientes.set_index("n")["Ex_quimica"].to_dict()
fluido_corriente=df_speco_corrientes.set_index("n")["fluido"].to_dict()
subunidad_corriente=df_speco_corrientes.set_index("n")["subunidad"].to_dict()
bloque_corriente=df_speco_corrientes.set_index("n")["bloque"].to_dict()
variable_costo_corriente=df_speco_corrientes.set_index("n")["variable_costo"].to_dict()

print(df_speco_corrientes[["n","bloque","subunidad","Ex","Ex_fisica","Ex_quimica","variable_costo"]].to_string(index=False))
print()
print("Corrientes SPECO:",len(df_speco_corrientes))
print("Con exergia quimica:",int(df_speco_corrientes["usa_ex_quimica"].sum()))
```

```python
def _valor_parametro(p,claves,default=np.nan):
    for k in claves:
        if k in p:
            try:
                return float(p[k])
            except:
                pass
    return default

def _Q_subunidad(df_QW,nombres):
    for nombre in nombres:
        v=df_QW.loc[df_QW["Subunidad"]==nombre,"Q(kW)"]
        if len(v):
            return float(v.iloc[0])
    return np.nan

def _W_subunidad(df_QW,nombres):
    for nombre in nombres:
        v=df_QW.loc[df_QW["Subunidad"]==nombre,"W(kW)"]
        if len(v):
            return float(v.iloc[0])
    return np.nan

def _crf(i,N):
    i=float(i)
    N=float(N)
    if abs(i)<=1e-12:
        return 1.0/N
    return i*(1.0+i)**N/((1.0+i)**N-1.0)

fronteras_speco={
    "cQsolar_util":0.0,
    "c3":0.0,
    "c4":0.0,
    "c11":0.0,
    "c28":0.0,
    "c38":0.0,
    "c43":0.0,
    "c71":0.0,
    "c73":0.0,
    "c75":0.0
}

param_econ_speco={}
param_econ_speco["phi"]=_valor_parametro(parametros_base,["phi"],1.06)
param_econ_speco["t_operacion_anual"]=_valor_parametro(parametros_base,["t_operacion_anual","t"],7446.0)
param_econ_speco["i"]=_valor_parametro(parametros_base,["i"],0.10)
param_econ_speco["N"]=_valor_parametro(parametros_base,["N"],20.0)
param_econ_speco["CRF"]=_crf(param_econ_speco["i"],param_econ_speco["N"])

vars_aux_speco={}

vars_aux_speco["W_comp_efgt"]=abs(_W_subunidad(df_QW,["Compresor(11->12)"]))
vars_aux_speco["W_tg"]=abs(_W_subunidad(df_QW,["Turbina_gas(14->15)"]))
vars_aux_speco["Wnet_EFGT"]=vars_aux_speco["W_tg"]-vars_aux_speco["W_comp_efgt"]

vars_aux_speco["W_turb_src"]=abs(_W_subunidad(df_QW,["Turbina_SRC(21->22)"]))
vars_aux_speco["W_bomba_src"]=abs(_W_subunidad(df_QW,["Bomba_SRC(23->24)"]))
vars_aux_speco["Wnet_SRC"]=vars_aux_speco["W_turb_src"]-vars_aux_speco["W_bomba_src"]

vars_aux_speco["W_turb_orc"]=abs(_W_subunidad(df_QW,["Turbina_ORC(31->32)"]))
vars_aux_speco["W_bomba_orc"]=abs(_W_subunidad(df_QW,["Bomba_ORC(33->34)"]))
vars_aux_speco["Wnet_ORC"]=vars_aux_speco["W_turb_orc"]-vars_aux_speco["W_bomba_orc"]

vars_aux_speco["W_bomba_seac"]=abs(_W_subunidad(df_QW,["Bomba_SEAC(61->62)"]))

eta_gen_efgt=_valor_parametro(parametros_base,["eta_generador_efgt"],1.0)
eta_gen_src=_valor_parametro(parametros_base,["eta_generador_src"],1.0)
eta_gen_orc=_valor_parametro(parametros_base,["eta_generador_orc"],1.0)

vars_aux_speco["W_elec_efgt"]=vars_aux_speco["Wnet_EFGT"]*eta_gen_efgt
vars_aux_speco["W_elec_src"]=vars_aux_speco["Wnet_SRC"]*eta_gen_src
vars_aux_speco["W_elec_orc"]=vars_aux_speco["Wnet_ORC"]*eta_gen_orc

vars_aux_speco["Qsolar_util"]=_valor_parametro(parametros_base,["Qsolar_util"],np.nan)
vars_aux_speco["Q_generador_seac"]=_valor_parametro(parametros_base,["Qg","Q_generador_seac"],np.nan)

df_fronteras_speco=pd.DataFrame({
    "variable":list(fronteras_speco.keys()),
    "costo_especifico":list(fronteras_speco.values())
}).sort_values("variable").reset_index(drop=True)

df_param_econ_speco=pd.DataFrame({
    "parametro":list(param_econ_speco.keys()),
    "valor":list(param_econ_speco.values())
})

df_vars_aux_speco=pd.DataFrame({
    "variable":list(vars_aux_speco.keys()),
    "valor":list(vars_aux_speco.values())
}).sort_values("variable").reset_index(drop=True)

Z_CI_speco={eq:np.nan for eq in df_speco_equipos["equipo"]}
Z_dot_speco={eq:np.nan for eq in df_speco_equipos["equipo"]}

print(df_fronteras_speco.to_string(index=False))
print()
print(df_param_econ_speco.to_string(index=False))
print()
print(df_vars_aux_speco.to_string(index=False))
```

```python
def _corriente_obj(n):
    if n in c_total:
        return c_total[n]
    raise KeyError(f"No existe la corriente {n} en c_total")

def _prop_corriente(n,nombres,default=np.nan):
    co=_corriente_obj(n)
    for nombre in nombres:
        if hasattr(co,nombre):
            try:
                return float(getattr(co,nombre))
            except:
                pass
        if isinstance(co,dict) and nombre in co:
            try:
                return float(co[nombre])
            except:
                pass
    return default

def _Tcorr(n):
    return _prop_corriente(n,["T","T_C"],np.nan)

def _dTlm(dt1,dt2):
    dt1=float(dt1)
    dt2=float(dt2)
    if pd.isna(dt1) or pd.isna(dt2):
        return np.nan
    if abs(dt1)<=1e-12 or abs(dt2)<=1e-12:
        return np.nan
    if dt1*dt2<=0:
        return np.nan
    if abs(dt1-dt2)<=1e-12:
        return abs(dt1)
    return abs((dt1-dt2)/np.log(dt1/dt2))

def _Q(df_QW,nombres):
    for nombre in nombres:
        v=df_QW.loc[df_QW["Subunidad"]==nombre,"Q(kW)"]
        if len(v):
            return abs(float(v.iloc[0]))
    return np.nan

def _A_hx(Q,U,dt1,dt2):
    dTlm=_dTlm(dt1,dt2)
    if pd.isna(Q) or pd.isna(U) or pd.isna(dTlm):
        return np.nan
    if abs(U)<=1e-12 or abs(dTlm)<=1e-12:
        return np.nan
    return Q/(U*dTlm)

def _Z_hx(A):
    if pd.isna(A):
        return np.nan
    return 130.0*((A/0.093)**0.78)

def _Z_ptc(A):
    if pd.isna(A):
        return np.nan
    return 13925.0*(A**0.17)

def _Z_potencia(W,factor):
    if pd.isna(W):
        return np.nan
    return factor*abs(W)

def _Z_capacidad(x,factor):
    if pd.isna(x):
        return np.nan
    return factor*abs(x)

U_speco={}
U_speco["SPH"]=0.25
U_speco["HTHE"]=0.125
U_speco["GV_SRC"]=0.20
U_speco["COND_DHU"]=0.25
U_speco["GV_ORC"]=0.25
U_speco["COND_ORC"]=0.25
U_speco["CAL_MED"]=0.25

A_speco={}
Q_speco={}
dt_speco={}

A_speco["PTC"]=_valor_parametro(parametros_base,["A_ptc","A_PTC","A_colector","Collector_area"],284.0)

Q_speco["SPH"]=_Q(df_QW,["Precalentador_solar(12->13)"])
dt_speco["SPH"]={}
dt_speco["SPH"]["dt1"]=_Tcorr(1)-_Tcorr(13)
dt_speco["SPH"]["dt2"]=_Tcorr(2)-_Tcorr(12)
A_speco["SPH"]=_A_hx(Q_speco["SPH"],U_speco["SPH"],dt_speco["SPH"]["dt1"],dt_speco["SPH"]["dt2"])

Q_speco["HTHE"]=_Q(df_QW,["HTHE_altaT(13->14)"])
dt_speco["HTHE"]={}
dt_speco["HTHE"]["dt1"]=_Tcorr(5)-_Tcorr(14)
dt_speco["HTHE"]["dt2"]=_Tcorr(6)-_Tcorr(13)
A_speco["HTHE"]=_A_hx(Q_speco["HTHE"],U_speco["HTHE"],dt_speco["HTHE"]["dt1"],dt_speco["HTHE"]["dt2"])

Q_speco["GV_SRC"]=_Q(df_QW,["Generador_SRC(15->16)"])
dt_speco["GV_SRC"]={}
dt_speco["GV_SRC"]["dt1"]=_Tcorr(15)-_Tcorr(21)
dt_speco["GV_SRC"]["dt2"]=_Tcorr(16)-_Tcorr(24)
A_speco["GV_SRC"]=_A_hx(Q_speco["GV_SRC"],U_speco["GV_SRC"],dt_speco["GV_SRC"]["dt1"],dt_speco["GV_SRC"]["dt2"])

Q_speco["COND_DHU"]=_Q(df_QW,["DHU(28->29)"])
dt_speco["COND_DHU"]={}
dt_speco["COND_DHU"]["dt1"]=_Tcorr(22)-_Tcorr(29)
dt_speco["COND_DHU"]["dt2"]=_Tcorr(23)-_Tcorr(28)
A_speco["COND_DHU"]=_A_hx(Q_speco["COND_DHU"],U_speco["COND_DHU"],dt_speco["COND_DHU"]["dt1"],dt_speco["COND_DHU"]["dt2"])

Q_speco["GV_ORC"]=_Q(df_QW,["Evap_ORC(34->31)"])
dt_speco["GV_ORC"]={}
dt_speco["GV_ORC"]["dt1"]=_Tcorr(16)-_Tcorr(31)
dt_speco["GV_ORC"]["dt2"]=_Tcorr(17)-_Tcorr(34)
A_speco["GV_ORC"]=_A_hx(Q_speco["GV_ORC"],U_speco["GV_ORC"],dt_speco["GV_ORC"]["dt1"],dt_speco["GV_ORC"]["dt2"])

Q_speco["COND_ORC"]=_Q(df_QW,["Cond_ORC(32->33)"])
dt_speco["COND_ORC"]={}
dt_speco["COND_ORC"]["dt1"]=_Tcorr(32)-_Tcorr(39)
dt_speco["COND_ORC"]["dt2"]=_Tcorr(33)-_Tcorr(38)
A_speco["COND_ORC"]=_A_hx(Q_speco["COND_ORC"],U_speco["COND_ORC"],dt_speco["COND_ORC"]["dt1"],dt_speco["COND_ORC"]["dt2"])

Q_speco["CAL_MED"]=_Q(df_QW,["Calentador_MED(6->7)","Calentador_agua_MED(41->42)"])
dt_speco["CAL_MED"]={}
dt_speco["CAL_MED"]["dt1"]=_Tcorr(6)-_Tcorr(42)
dt_speco["CAL_MED"]["dt2"]=_Tcorr(7)-_Tcorr(41)
A_speco["CAL_MED"]=_A_hx(Q_speco["CAL_MED"],U_speco["CAL_MED"],dt_speco["CAL_MED"]["dt1"],dt_speco["CAL_MED"]["dt2"])

for eq in df_speco_equipos["equipo"]:
    Z_CI_speco[eq]=np.nan
    Z_dot_speco[eq]=np.nan

Z_CI_speco["PTC"]=_Z_ptc(A_speco["PTC"])
Z_CI_speco["SPH"]=_Z_hx(A_speco["SPH"])
Z_CI_speco["HTHE"]=_Z_hx(A_speco["HTHE"])
Z_CI_speco["GV_SRC"]=_Z_hx(A_speco["GV_SRC"])
Z_CI_speco["COND_DHU"]=_Z_hx(A_speco["COND_DHU"])
Z_CI_speco["GV_ORC"]=_Z_hx(A_speco["GV_ORC"])
Z_CI_speco["COND_ORC"]=_Z_hx(A_speco["COND_ORC"])
Z_CI_speco["CAL_MED"]=_Z_hx(A_speco["CAL_MED"])

m3=_valor_parametro(parametros_base,["m3"],np.nan)
pci_b=_valor_parametro(parametros_base,["PCI_bagazo","PCI_bagazo_kJ_kg"],np.nan)
if pd.notna(m3) and pd.notna(pci_b):
    Z_CI_speco["BCC"]=1000.0*m3*(pci_b/1000.0)

Z_CI_speco["COMP_EFGT"]=_Z_potencia(vars_aux_speco["W_comp_efgt"],900.0)
Z_CI_speco["TG"]=_Z_potencia(vars_aux_speco["W_elec_efgt"],950.0)

Z_CI_speco["TURB_SRC"]=_Z_potencia(vars_aux_speco["W_elec_src"],850.0)
Z_CI_speco["BOMBA_SRC"]=_Z_potencia(vars_aux_speco["W_bomba_src"],400.0)

Z_CI_speco["TURB_ORC"]=_Z_potencia(vars_aux_speco["W_elec_orc"],850.0)
Z_CI_speco["BOMBA_ORC"]=_Z_potencia(vars_aux_speco["W_bomba_orc"],400.0)

Z_CI_speco["MED"]=_Z_capacidad(_valor_parametro(parametros_base,["m59"],np.nan),500000.0)
Z_CI_speco["SEAC"]=_Z_capacidad(_valor_parametro(parametros_base,["Q_evaporador_seac","Q_ev_seac"],np.nan),1800.0)

for eq,val in Z_CI_speco.items():
    if pd.notna(val):
        Z_dot_speco[eq]=val*param_econ_speco["CRF"]*param_econ_speco["phi"]/param_econ_speco["t_operacion_anual"]

df_A_speco=pd.DataFrame({
    "equipo":["PTC","SPH","HTHE","GV_SRC","COND_DHU","GV_ORC","COND_ORC","CAL_MED"],
    "A":[
        A_speco.get("PTC",np.nan),
        A_speco.get("SPH",np.nan),
        A_speco.get("HTHE",np.nan),
        A_speco.get("GV_SRC",np.nan),
        A_speco.get("COND_DHU",np.nan),
        A_speco.get("GV_ORC",np.nan),
        A_speco.get("COND_ORC",np.nan),
        A_speco.get("CAL_MED",np.nan)
    ]
})

df_Z_speco=pd.DataFrame({
    "equipo":list(Z_CI_speco.keys()),
    "Z_CI":[Z_CI_speco[k] for k in Z_CI_speco.keys()],
    "Z_dot":[Z_dot_speco[k] for k in Z_dot_speco.keys()]
})

print(df_A_speco.to_string(index=False))
print()
print(df_Z_speco.to_string(index=False))
```

6. Sistema de ecuaciones SPECO

```python
def _variables_ec(df_ec):
    vars_ec=[]
    for _,row in df_ec.iterrows():
        for t in row["lhs"]:
            vars_ec.append(t["var"])
        for t in row["rhs"]:
            vars_ec.append(t["var"])
    return sorted(set(vars_ec))

def _texto_ec(row):
    def _lado(ts):
        if not ts:
            return "0"
        return " + ".join([f"{t['coef']:.12g}*{t['var']}" for t in ts])
    return _lado(row["lhs"])+" = "+_lado(row["rhs"])+" + "+f"{row['constante']:.12g}"

variables_speco=_variables_ec(df_ec_speco)
idx_var={v:i for i,v in enumerate(variables_speco)}

n_ec=len(df_ec_speco)
n_var=len(variables_speco)

A_speco=np.zeros((n_ec,n_var))
B_speco=np.zeros(n_ec)

filas_info=[]

for i,(_,row) in enumerate(df_ec_speco.iterrows()):
    for t in row["lhs"]:
        A_speco[i,idx_var[t["var"]]]+=t["coef"]
    for t in row["rhs"]:
        A_speco[i,idx_var[t["var"]]]-=t["coef"]
    B_speco[i]=row["constante"]

    filas_info.append({
        "ecuacion":i,
        "equipo":row["equipo"],
        "tipo_ecuacion":row["tipo_ecuacion"],
        "ecuacion_texto":_texto_ec(row)
    })

df_info_speco=pd.DataFrame(filas_info)

rango_A=np.linalg.matrix_rank(A_speco)
rango_AB=np.linalg.matrix_rank(np.column_stack([A_speco,B_speco]))

if rango_A<rango_AB:
    diagnostico_speco="incompatible"
elif rango_A==rango_AB and rango_A<n_var:
    diagnostico_speco="infinitas_soluciones"
elif rango_A==rango_AB and rango_A==n_var:
    diagnostico_speco="solucion_unica"
else:
    diagnostico_speco="indeterminado"

apariciones=np.sum(np.abs(A_speco)>1e-12,axis=0)

df_diag_speco=pd.DataFrame([
    {"indicador":"n_ecuaciones","valor":n_ec},
    {"indicador":"n_incognitas","valor":n_var},
    {"indicador":"rango_A","valor":rango_A},
    {"indicador":"rango_AB","valor":rango_AB},
    {"indicador":"diagnostico","valor":diagnostico_speco},
    {"indicador":"dimension_A","valor":f"{A_speco.shape[0]}x{A_speco.shape[1]}"}
])

df_vars_speco=pd.DataFrame({
    "variable":variables_speco,
    "apariciones":apariciones.astype(int)
}).sort_values(["apariciones","variable"],ascending=[True,True]).reset_index(drop=True)

df_vars_flojas_speco=df_vars_speco[df_vars_speco["apariciones"]<=1].reset_index(drop=True)
df_vars_poco_acopladas_speco=df_vars_speco[df_vars_speco["apariciones"]<=2].reset_index(drop=True)

print(df_diag_speco.to_string(index=False))
print()
print(df_vars_flojas_speco.to_string(index=False))
print()
print(df_vars_poco_acopladas_speco.to_string(index=False))
```

```python
lista_equipos=df_speco_equipos["equipo"].tolist()

Z_dot_vector=[]
C_D_vector=[]
Ex_D_vector=[]
c_F_vector=[]

for _,row in df_speco_equipos.iterrows():
    eq=row["equipo"]

    Ex_in=0.0 #kW
    for n in row["corrientes_entrada"]:
        Ex_in+=ex_corriente.get(n,0.0)

    if row["trabajo_in"]!="":
        Ex_in+=vars_aux_speco[row["trabajo_in"]] #kW

    if row["calor_in"]!="":
        Ex_in+=coef_wq_speco[row["calor_in"]] #kW

    Ex_out=0.0 #kW
    for n in row["corrientes_salida"]:
        Ex_out+=ex_corriente.get(n,0.0)

    if row["trabajo_out"]!="":
        Ex_out+=vars_aux_speco[row["trabajo_out"]] #kW

    if row["calor_out"]!="":
        Ex_out+=coef_wq_speco[row["calor_out"]] #kW

    Ex_D=max(Ex_in-Ex_out,0.0) #kW

    costos_entrada=[]

    for n in row["corrientes_entrada"]:
        var=variable_costo_corriente[n]
        if var in sol_speco:
            costos_entrada.append(sol_speco[var])

    if row["trabajo_in"]!="":
        var_w="c"+row["trabajo_in"]
        if var_w in sol_speco:
            costos_entrada.append(sol_speco[var_w])

    if row["calor_in"]!="":
        var_q="c"+row["calor_in"]
        if var_q in sol_speco:
            costos_entrada.append(sol_speco[var_q])

    if len(costos_entrada)>0:
        c_F=float(np.mean(costos_entrada)) #$/kWh_ex
    else:
        c_F=0.0 #$/kWh_ex

    C_D=c_F*Ex_D #$/h, si tu c está referido a kWh_ex y Ex a kW

    if eq in Z_dot_speco and pd.notna(Z_dot_speco[eq]):
        Z_dot=float(Z_dot_speco[eq]) #$/h
    else:
        Z_dot=0.0 #$/h

    Z_dot_vector.append(Z_dot)
    C_D_vector.append(C_D)
    Ex_D_vector.append(Ex_D)
    c_F_vector.append(c_F)

df_exergoeco=pd.DataFrame({
    "equipo":lista_equipos,
    "Ex_D (kW)":Ex_D_vector,
    "c_F ($/kWh_ex)":c_F_vector,
    "C_D ($/h)":C_D_vector,
    "Z_dot ($/h)":Z_dot_vector
})

df_exergoeco["f (-)"]=df_exergoeco["Z_dot ($/h)"]/(df_exergoeco["Z_dot ($/h)"]+df_exergoeco["C_D ($/h)"]+1e-12)
df_exergoeco["Z_frac (-)"]=df_exergoeco["f (-)"]
df_exergoeco["D_frac (-)"]=1.0-df_exergoeco["f (-)"]

print(df_exergoeco.to_string(index=False))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey

df_exergoeco_plot=df_exergoeco.copy()

df_exergoeco_plot["total ($/h)"]=df_exergoeco_plot["Z_dot ($/h)"]+df_exergoeco_plot["C_D ($/h)"]
df_exergoeco_plot["Z_frac (-)"]=np.where(df_exergoeco_plot["total ($/h)"]>0,df_exergoeco_plot["Z_dot ($/h)"]/df_exergoeco_plot["total ($/h)"],0.0)
df_exergoeco_plot["D_frac (-)"]=np.where(df_exergoeco_plot["total ($/h)"]>0,df_exergoeco_plot["C_D ($/h)"]/df_exergoeco_plot["total ($/h)"],0.0)

clasif_prod={
    "producto_DHU":"Calefaccion",
    "producto_agua":"Agua desalada",
    "producto_frio":"Frio"
}

def graficar_factor_exergoeconomico(df):
    dfp=df.sort_values("Z_frac (-)").reset_index(drop=True)
    y=range(len(dfp))
    plt.figure(figsize=(9,7))
    plt.barh(y,dfp["Z_frac (-)"],label="Costo por inversion")
    plt.barh(y,dfp["D_frac (-)"],left=dfp["Z_frac (-)"],label="Costo por irreversibilidades")
    plt.yticks(list(y),dfp["equipo"])
    plt.xlim(0,1)
    plt.xlabel("Fraccion (-)")
    plt.ylabel("Equipo (-)")
    plt.legend()
    plt.grid(axis="x",linestyle="--",alpha=0.4)
    plt.tight_layout()
    plt.show()

def graficar_cd_vs_z(df):
    plt.figure(figsize=(7,6))
    plt.scatter(df["Z_dot ($/h)"],df["C_D ($/h)"])
    for _,r in df.iterrows():
        plt.annotate(r["equipo"],(r["Z_dot ($/h)"],r["C_D ($/h)"]),xytext=(4,4),textcoords="offset points",fontsize=8)
    plt.xlabel("Z_dot ($/h)")
    plt.ylabel("C_D ($/h)")
    plt.grid(linestyle="--",alpha=0.4)
    plt.tight_layout()
    plt.show()

def graficar_exd_vs_cd(df):
    plt.figure(figsize=(7,6))
    plt.scatter(df["Ex_D (kW)"],df["C_D ($/h)"])
    for _,r in df.iterrows():
        plt.annotate(r["equipo"],(r["Ex_D (kW)"],r["C_D ($/h)"]),xytext=(4,4),textcoords="offset points",fontsize=8)
    plt.xlabel("Ex_D (kW)")
    plt.ylabel("C_D ($/h)")
    plt.grid(linestyle="--",alpha=0.4)
    plt.tight_layout()
    plt.show()

def graficar_costos_productos(df):
    dfp=df[df["clasificacion (-)"].isin(clasif_prod.keys())].copy()
    dfp["producto (-)"]=dfp["clasificacion (-)"].map(clasif_prod)
    plt.figure(figsize=(7,5))
    plt.bar(dfp["producto (-)"],dfp["c ($/kWh_ex)"])
    plt.ylabel("c ($/kWh_ex)")
    plt.xlabel("Producto (-)")
    plt.grid(axis="y",linestyle="--",alpha=0.4)
    plt.tight_layout()
    plt.show()

def graficar_flujo_costo_productos(df):
    dfp=df[df["clasificacion (-)"].isin(clasif_prod.keys())].copy()
    dfp["producto (-)"]=dfp["clasificacion (-)"].map(clasif_prod)
    plt.figure(figsize=(7,5))
    plt.bar(dfp["producto (-)"],dfp["Ċ ($/h)"])
    plt.ylabel("Ċ ($/h)")
    plt.xlabel("Producto (-)")
    plt.grid(axis="y",linestyle="--",alpha=0.4)
    plt.tight_layout()
    plt.show()

def graficar_perfil_costos_corrientes(df):
    dfp=df.copy().sort_values("n (-)").reset_index(drop=True)
    plt.figure(figsize=(12,5))
    plt.plot(dfp["n (-)"],dfp["c ($/kWh_ex)"],marker="o")
    plt.xlabel("Corriente n (-)")
    plt.ylabel("c ($/kWh_ex)")
    plt.grid(linestyle="--",alpha=0.4)
    plt.tight_layout()
    plt.show()

def graficar_perfil_flujo_costo_corrientes(df):
    dfp=df.copy().sort_values("n (-)").reset_index(drop=True)
    plt.figure(figsize=(12,5))
    plt.bar(dfp["n (-)"].astype(str),dfp["Ċ ($/h)"])
    plt.xlabel("Corriente n (-)")
    plt.ylabel("Ċ ($/h)")
    plt.xticks(rotation=90)
    plt.grid(axis="y",linestyle="--",alpha=0.4)
    plt.tight_layout()
    plt.show()

def graficar_mapa_calor(df):
    dft=df[["equipo","Ex_D (kW)","c_F ($/kWh_ex)","C_D ($/h)","Z_dot ($/h)","f (-)"]].copy().set_index("equipo")
    mat=dft.values
    fig,ax=plt.subplots(figsize=(8,7))
    im=ax.imshow(mat,aspect="auto")
    ax.set_yticks(np.arange(len(dft.index)))
    ax.set_yticklabels(dft.index)
    ax.set_xticks(np.arange(len(dft.columns)))
    ax.set_xticklabels(dft.columns,rotation=45,ha="right")
    for i in range(mat.shape[0]):
        for j in range(mat.shape[1]):
            ax.text(j,i,f"{mat[i,j]:.3g}",ha="center",va="center",fontsize=7)
    plt.colorbar(im)
    plt.tight_layout()
    plt.show()

def graficar_contribucion_cd(df):
    dft=df.sort_values("C_D ($/h)",ascending=False).reset_index(drop=True)
    plt.figure(figsize=(9,5))
    plt.bar(dft["equipo"],dft["C_D ($/h)"])
    plt.ylabel("C_D ($/h)")
    plt.xlabel("Equipo (-)")
    plt.xticks(rotation=90)
    plt.grid(axis="y",linestyle="--",alpha=0.4)
    plt.tight_layout()
    plt.show()

def graficar_contribucion_z(df):
    dft=df.sort_values("Z_dot ($/h)",ascending=False).reset_index(drop=True)
    plt.figure(figsize=(9,5))
    plt.bar(dft["equipo"],dft["Z_dot ($/h)"])
    plt.ylabel("Z_dot ($/h)")
    plt.xlabel("Equipo (-)")
    plt.xticks(rotation=90)
    plt.grid(axis="y",linestyle="--",alpha=0.4)
    plt.tight_layout()
    plt.show()

def graficar_pastel_cd(df):
    dft=df[df["C_D ($/h)"]>0].copy().sort_values("C_D ($/h)",ascending=False)
    plt.figure(figsize=(7,7))
    plt.pie(dft["C_D ($/h)"],labels=dft["equipo"],autopct="%1.1f%%")
    plt.tight_layout()
    plt.show()

def graficar_pastel_total(df):
    dft=df[df["total ($/h)"]>0].copy().sort_values("total ($/h)",ascending=False)
    plt.figure(figsize=(7,7))
    plt.pie(dft["total ($/h)"],labels=dft["equipo"],autopct="%1.1f%%")
    plt.tight_layout()
    plt.show()

def graficar_sankey_productos(df):
    dfp=df[df["clasificacion (-)"].isin(clasif_prod.keys())].copy()
    Q_solar=float(df_trabajos_calores_speco.loc[df_trabajos_calores_speco["variable (-)"]=="cQsolar_util","c ($/kWh_ex)"].iloc[0])*float(df_ExWQ_speco.loc[df_ExWQ_speco["variable"]=="Qsolar_util","Ex_asociada"].iloc[0]) #$/h
    c3=float(sol_speco.get("c3",0.0))
    Ex3=float(ex_corriente.get(3,0.0)) #kW
    C_biomasa=c3*Ex3 #$/h
    C_elec=float(sol_speco.get("cW_elec_efgt",0.0))*float(vars_aux_speco.get("W_elec_efgt",0.0))
    C_elec+=float(sol_speco.get("cW_elec_src",0.0))*float(vars_aux_speco.get("W_elec_src",0.0))
    C_elec+=float(sol_speco.get("cW_elec_orc",0.0))*float(vars_aux_speco.get("W_elec_orc",0.0))
    C_cal=float(dfp.loc[dfp["clasificacion (-)"]=="producto_DHU","Ċ ($/h)"].sum())
    C_agua=float(dfp.loc[dfp["clasificacion (-)"]=="producto_agua","Ċ ($/h)"].sum())
    C_frio=float(dfp.loc[dfp["clasificacion (-)"]=="producto_frio","Ċ ($/h)"].sum())
    C_res=np.maximum(Q_solar+C_biomasa-(C_elec+C_cal+C_agua+C_frio),0.0)

    plt.figure(figsize=(10,6))
    sankey=Sankey(unit=None,format="%.2f")
    sankey.add(
        flows=[Q_solar,C_biomasa,-C_elec,-C_cal,-C_agua,-C_frio,-C_res],
        labels=["Solar","Biomasa","Electricidad","Calefaccion","Agua","Frio","Residuo"],
        orientations=[1,1,-1,-1,-1,-1,0]
    )
    sankey.finish()
    plt.tight_layout()
    plt.show()

graficar_factor_exergoeconomico(df_exergoeco_plot)
graficar_cd_vs_z(df_exergoeco_plot)
graficar_exd_vs_cd(df_exergoeco_plot)
graficar_costos_productos(df_tabla_final_speco)
graficar_flujo_costo_productos(df_tabla_final_speco)
graficar_perfil_costos_corrientes(df_tabla_final_speco)
graficar_perfil_flujo_costo_corrientes(df_tabla_final_speco)
graficar_mapa_calor(df_exergoeco_plot)
graficar_contribucion_cd(df_exergoeco_plot)
graficar_contribucion_z(df_exergoeco_plot)
graficar_pastel_cd(df_exergoeco_plot)
graficar_pastel_total(df_exergoeco_plot)
graficar_sankey_productos(df_tabla_final_speco)
```
